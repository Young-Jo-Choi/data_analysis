# 데이터 합치기
방법:
- pandas.merge : 하나 이상의 키를 기준으로 
- pandas.concat : 하나의 축을 따라 이어붙임
- pandas.combile_first : 두 객체를 포갬(누락된 데이터를 다른 객체의 데이터로 채움)

## 데이터베이스 스타일로 DataFrame 합치기
병합(merge)이나 조인(join) 연산은 관계형 데이터베이스의 핵심적 기능인데, 하나 이상의 키를 사용해서 데이터 집합의 로우를 합친다.

### 다대일의 경우
데카르트곱의 형태로 생각해도 된다.

```python
# df1 데이터는 key컬럼에 여러 개의 a,b 로우를 가지고 있고 df2의 데이터는 key컬럼에 유일한 로우를 가지고 있다.

df1 = pd.DataFrame({'key':['b','b','a','c','a','a','b'],'data1':range(7)})
df2 = pd.DataFrame({'key':['a','b','d'],'data2':range(3)})

pd.merge(df1,df2)    # key 컬럼 중 겹치지 않는 c,d의 데이터는 사라짐, 이 경우도 인자를 따로 줌으로서 포함시킬 수 있음. 아래 참조

# merge함수는 기본적으로 중복된 컬럼 이름을 키로 사용하는데 복잡한 데이터의 경우 에러가 날 수 있으니
# pd.merge(df1,df2,on='key') 처럼 on 옵션을 통해 키를 명시해주는 것이 좋다.
```

```python
# 중복된 컬럼 이름이 없는 경우 직접 지정해주면 된다.

df3 = pd.DataFrame({'key1':['b','b','a','c','a','a','b'],'data1':range(7)})
df4 = pd.DataFrame({'key2':['a','b','d'],'data2':range(3)}) 
pd.merge(df3,df4,left_on='key1',right_on='key2') 
```

```python
# how 옵션에 따라 조인 연산의 동작을 다르게 할 수 있다.

pd.merge(df1,df2,how='outer')  # 중복되지 않는 값까지 포함 
```

### how 옵션에 따른 조인 연산
- inner : 교집합(내부조인)
- outer : 합집합(외부조인)
- left/right : 왼쪽/오른쪽 테이블에 존재하는 모든 키 조합 사용
- output : 양쪽 테이블에 존재하는 모든 키 조합 사용

### 다대다 병합
직관적인 결과값은 아니지만 데카르트곱을 반환한다 생각하면 쉽다.

일대다와 다대다는 문법에 차이가 있는 것이 아니라 적용되는 데이터셋의 형태에 따른 분류이다.

```python
# 두 로우의 데카르트곱을 반환한다.

df1 = pd.DataFrame({'key':['b','b','a','c','a','b'],'data1':range(6)})
df2 = pd.DataFrame({'key':['a','b','a','b','d'],'data2':range(5)})
pd.merge(df1,df2,on='key',how='inner')
# 이 경우에도 how에 left, right를 줄 수 있고 마찬가지로 데카르트 곱의 기준을 삼는 것임
```

```python
#여러 개의 키를 병합하려면 컬럼 이름이 담긴 리스트를 넘긴다.

left = pd.DataFrame({'key1':['foo','foo','bar'],'key2':['one','two','one'],'lval':[1,2,3]})
right = pd.DataFrame({'key1':['foo','foo','bar','bar'],'key2':['one','one','one','two'],'rval':[4,5,6,7]})

pd.merge(left,right,on=['key1','key2'],how='outer')
```

```python
# suffixes 인자로 두 DataFrame 객체에서 겹치는 컬럼 이름 뒤에 붙일 문자열 지정해줄 수 있다.
# 아래 예시에서 key2는 겹치는 이름이지만 기준 키가 아니기 때문에 병합할 경우 똑같은 이름의 컬럼이 생길 수 있다. 이런 경우 사용 

pd.merge(left,right,on='key1',suffixes=('_left','_right'))
```

### merge 함수 인자 목록
- left/right : 병합하려는 DataFrame 중 왼쪽/오른쪽에 위치한 DataFrame
- left_on/right_on : 조인키로 사용할 왼쪽/오른쪽 DataFrame의 컬럼
- left_index/right_index : 조인키로 사용할 왼쪽/오른쪽 DataFrame의 색인 로우(다중 색인일 경우 키)
- sort : 조인키에 따라 병합된 데이터를 사전순으로 정렬, 기본값은 True, 대용량 데이터의 경우 False로 해놓으면 성능상의 이득 얻을 수 있음
- copy : False일 경우, 예외적인 경우에 데이터가 결과로 복사되지 않게 한다.기본값은 항상 복사가 이루어짐.
- indicator : merge라는 이름의 컬럼을 추가하여 각 로우의 소스가 어디인지 나타낸다.'left_only','right_only','both'의 값 가짐
- how, on, suffixes

## 색인 병합하기
위에 써있는 left/right_index = True 이용

혹은 join 메서드 이용

```python
left1 = pd.DataFrame({'key':['a','b','a','a','b','c'],'value':range(6)})
right1 = pd.DataFrame({'group_val':[3.5, 7]},index = ['a','b'])
print(left1)
print('\n')
print(right1)
```

```python
# 왼쪽 데이터의 key컬럼 값들과 오른쪽 데이터에서는 인덱스를 기준으로 병합

pd.merge(left1,right1,left_on='key',right_index = True)
```

```python
# 계층 색인된 데이터는 암묵적으로 여러 키를 병합하는 것이라 약간 복잡하다.

lefth = pd.DataFrame({'key1':['Ohio','Ohio','Ohio','Nevada','Nevada'],
                      'key2':[2000,2001,2002,2001,2002],'data':np.arange(5.)})
righth = pd.DataFrame(np.arange(12).reshape((6,2)),index=[['Nevada','Nevada','Ohio','Ohio','Ohio','Ohio'],
                                                         [2001,2000,2000,2000,2001,2002]],
                     columns=['event1','event2'])
print(lefth)
print('\n')
print(righth)
```

```python
pd.merge(lefth,righth,left_on=['key1','key2'],right_index=True, how='outer')
```

```python
# 양쪽에 공통적으로 존재하는 색인을 병합할 경우 left_index=True, right_index=True를 해줌

left2 = pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],index=['a','c','e'],columns=['Ohio','Nevada'])
right2 = pd.DataFrame([[7.,8.],[9.,10.],[11.,12.,],[13,14]],
                    index=['b','c','d','e'],columns=['Missouri','Alabama'])
pd.merge(left2,right2,how='outer',left_index=True,right_index=True)
```

join

```python
# 색인으로 병합할 때는 join 메서드가 편함(위와 똑같은 결과)
# join 메서드는 pd.join 같은 형식으로는 사용하지 못함

left2.join(right2,how='outer')
```

```python
# 색인 대 색인으로 DataFrame들을 병합-->그냥 간단히 DataFrame의 리스트를 join 메서드로 넘김

left2 = pd.DataFrame([[1.,2.],[3.,4.],[5.,6.]],index=['a','c','e'],columns=['Ohio','Nevada'])
right2 = pd.DataFrame([[7.,8.],[9.,10.],[11.,12.,],[13,14]],
                    index=['b','c','d','e'],columns=['Missouri','Alabama'])
another = pd.DataFrame([[7.,8.],[9.,10.],[11.,12.],[16.,17.]],
                     index=['a','c','e','f'],columns=['New York','Oregon']) # 추가될 컬럼

# 내부조인으로서 left2의 a,c,e만을 기준으로 병합함
left2.join([right2,another])
```

## 축 따라 이어붙이기 : concatenation
NumPy는 ndarray를 이어붙이는 concatenate 함수 제공

```python
arr=np.arange(12).reshape((3,4))
np.concatenate([arr,arr],axis=1)
```

pandas 객체에서는 concat 함수를 사용할 수 있다.
- 두 객체 색인의 교집합 혹은 합집합을 반환할지 정할 수 있음 : join='inner' or 'outer'(기본적으로는 외부조인)
- 축을 설정할 수 있음 : axis
- 합쳐진 결과에서 합쳐지기 전 객체의 데이터를 구분할 수 있음 : keys인자 활용

```python
s1 = pd.Series([0,1],index=['a','b'])
s2 = pd.Series([2,3,4],index=['c','d','e'])
s3 = pd.Series([5,6],index=['f','g'])
pd.concat([s1,s2,s3])
```

```python
# 병합하려는 축을 직접 설정해 줄 수도 있다.

pd.concat([s1,s3],axis=1,join_axes=[['a','b','f','e']])
```

```python
# 병합전의 개별 Series를 구분할 수 없는 문제가 생기는데, 
# 이어붙인 축에 계층적 색인을 지정해 식별 가능하도록 할 수 있다.---> keys 인자 사용


result = pd.concat([s1,s2,s3],keys=['one','two','three'])
result
```

```python
result.unstack()
```

```python
# Series를 axis=1로 병합하면 keys는 DataFrame의 컬럼명들이 된다.

result = pd.concat([s1,s2,s3],keys=['one','two','three'],axis=1)
result
```

```python
# DataFrame간에도 똑같은 방식으로 적용가능함

df1 = pd.DataFrame(np.arange(6).reshape((3,2)),index=['a','b','c'],columns=['one','two'])
df2 = pd.DataFrame(5+np.arange(4).reshape((2,2)),index=['a','c'],columns=['three','four'])
pd.concat([df1,df2],axis=1,keys=['first','second'],sort=True)
```

리스트 대신 사전을 넘기면 사전의 키가 keys 옵션으로 사용된다.

pd.concat([df1,df2],axis=1,keys=['first','second'],sort=True)

==

pd.concat({'first':df1,'second':df2},axis=1)

```python
#DataFrame의 로우 색인이 분석에 필요한 데이터를 포함하고 있지 않은 경우

df1 = pd.DataFrame(np.random.randn(3,4),columns=['a','b','c','d'])
df2 = pd.DataFrame(np.random.randn(2,3),columns=['b','d','a'])

print(df1,'\n\n',df2)
```

```python
pd.concat([df1,df2],sort=True)      # sort는 여기서는 별 의미가 없으니 무시하시오.
```

```python
# ignore_index : index의 원래 이름을 무시, range로 새로운 색인을 생성

pd.concat([df1,df2],sort=True,ignore_index=True)
```

### concat 함수 인자
- objs : 이어붙일 pandas 객체(Series,DataFrame)의 사전이나 리스트. 필수인자
- join_axes : 합집합/교집합 수행하는 대신 다른 n-1 축으로 사용할 색인을 지정한다.
- keys : 이어붙일 객체나 축에 대한 계층 색인을 생성하는데 연관됨. 리스트, 배열, 튜플의 배열, 배열의 리스트 등을 넣을 수 있다.
- levels : 계층 색인 레벨로 사용할 색인을 지정. keys가 넘어온 경우 여러 개의 색인을 지정한다.
- names : keys, levels 다 있는 경우 생성된 계층 레벨을 위한 이름
- verify_integrity : 이어붙인 객체에 중복되는 축이 있는지 검사, 있으면 예외를 발생, 기본값은 False로서 중복을 허용
- axis, join, ignore_index

## 겹치는 데이터 합치기(포개기)
두 데이터셋의 색인이 일부나 전체 겹치는 경우. 벡터화된 if-else 구문인 np.where함수 이용

pandas 객체 고유의 combine_first 메서드 역시 살펴봄

```python
a = pd.Series([np.nan,2.5,np.nan,3.5,4.5,np.nan],index=['f','e','d','c','b','a'])
b = pd.Series(np.arange(len(a),dtype=np.float64),index=['f','e','d','c','d','a'])
b[-1] = np.nan
print(a,'\n\n', b)
```

```python
np.where(pd.isnull(a),b,a)
```

combine_first 메서드는 위와 동일한 연산을 제공하며 데이터 정렬 기능까지 제공한다.

```python
b[:-2].combine_first(a[2:])

# b[:-2]에 c,d,e,f의 값은 포함됨. a[:2]에는 a,b,c,d가 있음. 
# b[:-2]에 이미 있는 값은 b[:-2]를 따르고 없는 값은 a[2:]에서 참고한다는 의미
```

```python
# DataFrame에서 combine_first는 컬럼에 대해 같은 동작을 함

df1 = pd.DataFrame({'a':[1.,np.nan,5.,np.nan],
                   'b':[np.nan,2.,np.nan,6.],
                   'c':range(2,18,4)})
df2 = pd.DataFrame({'a':[5.,4.,np.nan,3.,7.],
                   'b':[np.nan,3.,4.,6.,8.]})
print(df1,'\n')
print(df2)
df1.combine_first(df2)

# df1을 기본으로 없는 값에 대해서만 df2를 참고
```

# 재형성과 피벗
## 계층적 색인으로 재형성하기

- stack : 데이터의 컬럼을 로우로 피벗(회전)
- unstack : 데이터의 로우를 컬럼으로 피벗

기본적으로 가장 안쪽에 있는 라벨(가장 높은 레벨)부터 끄집어내는데, 레벨 숫자나 이름을 전달해 끄집어낼 단계를 지정할 수 있다. 

결과물에서도 가장 안쪽 단계로 이동(예시로 확인합시다)

```python
data = pd.DataFrame(np.arange(6).reshape((2,3)), 
                    index=pd.Index(['Ohio','Colorado'],name='state'),       # frame.index.names=~~ 처럼 따로 빼낼 수도 있음
                    columns=pd.Index(['one','two','three'],name='number'))
data
```

```python
data.stack()   # 컬럼을 로우로 보냄(단층색인된 컬럼을 가진 DataFrame을 Series로 만듦)
```

```python
data.stack().unstack()   # 색인(로우)을 컬럼으로 보내는데 기본적으로 가장 안쪽 레벨(가장 큰 레벨)에서부터 끄집어냄
```

```python
data.stack().unstack(0)   # 0번 레벨인 state를 컬럼으로 보냄

# data.stack().unstack('state') --> 계층 이름을 직접 입력하는 것도 물론 가능
```

```python
# stack의 경우 dropna=False 헤주지 않으면 누락된 데이터를 자동으로 걸러냄

s1 = pd.Series([0,1,2,3],index=['a','b','c','d'])
s2 = pd.Series([4,5,6],index=['c','d','e'])
data2 = pd.concat([s1,s2],keys=['one','two'])

print('data2.unstack().stack():\n\n',data2.unstack().stack())
print('\ndata2.unstack().stack(dropna=False):\n\n',data2.unstack().stack(dropna=False))
```

## 긴 형식에서 넓은 형식으로 피벗하기
예제를 사용

```python
data = pd.read_csv('macrodata.csv')
data.head()
```

```python
periods = pd.PeriodIndex(year=data.loc[:,'year'], quarter=data.quarter, name='data')
columns = pd.Index(['realgdp','infl','unemp'],name='item')
data = data.reindex(columns=columns)
data.index = periods.to_timestamp('D','end')
ldata = data.stack().reset_index().rename(columns={0:'value'})
```

```python
ldata[:10]
```

```python
# pivot 함수의 처음 두 인자는 로우와 컬럼 색인으로 사용될 컬럼 이름이고 
# 마지막 두 인자는 DataFrame에 채워넣을 값을 담고 있는 컬럼 이름이다

pivoted = ldata.pivot('data','item','value')
pivoted
```

```python
# 한 번에 두 개의 컬럼을 동시에 변형한다고 하자.

ldata['value2'] = np.random.randn(len(ldata))
ldata[:10]
```

```python
# 마지막 인자를 생략해서 계층적 컬럼을 가지는 DataFrame을 얻을 수 있다.

pivoted = ldata.pivot('data','item')
pivoted[:5]
```

```python
pivoted['value'][:5]
```

```python
# pivot은 set_index를 사용해서 계층적 색인을 만들고 unstack 메서드를 이용해서 형태를 변경하는 단축키 같은 메서드이다.

unstacked = ldata.set_index(['data','item']).unstack('item')
unstacked == pivoted
```

## 넓은 형식에서 긴 형식으로 피벗하기
pivot과 반대되는 연산은 pandas.melt다. 

하나의 컬럼을 여러 개의 새로운 DataFrame으로 생성하기 보다는 여러 컬럼을 하나로 병합하고 DataFrame을 입력보다 긴 형태로 만들어낸다.

```python
df=pd.DataFrame({'key':['foo','bar','baz'],'A':[1,2,3],'B':[4,5,6],'C':[7,8,9]})
df
```

```python
# melt : 여러 컬럼을 하나로

melted = pd.melt(df,['key'])
melted       # 어떤 컬럼을 구분자로 사용할 것인지 지정해야함
```

```python
reshaped = melted.pivot('key','variable','value')
reshaped   
# pivot 함수의 처음 두 인자는 로우와 컬럼 색인으로 사용될 컬럼 이름이고 
# 마지막 두 인자는 DataFrame에 채워넣을 값을 담고 있는 컬럼 이름이다
```

```python
reshaped.reset_index()   # 로우의 계층적 색인을 컬럼으로 이동(key가 컬럼으로 이동됨)
                         # 컬럼name은 variable
```

```python
# 데이터 값으로 사용할 컬럼들의 집합을 지정할 수도 있다.

pd.melt(df,id_vars=['key'],value_vars=['A','B'])
```

```python
# 구분자 없이 사용할 수도 있다.

pd.melt(df,value_vars=['A','B','C'])
```

```python
pd.melt(df,value_vars=['A','key','B'])
```
