```python
import pandas as pd
import numpy as np
```

```python
starbucks = pd.read_excel('st_stat.xlsx')
pop = pd.read_excel('st_pop.xlsx')
work = pd.read_excel('st_work.xlsx')
```

```python
starbucks
```

```python
new = pd.merge(pop,work)
```

```python
new
```

```python
new2 = starbucks['시군구명']
```

```python
new3 = pd.DataFrame(new2.value_counts())
```

```python
new3.columns = ['매장수']
```

```python
new3
```

```python
result = pd.merge(new,new3,left_on = '시군구명',right_index=True)
```

```python
result
```

```python
y = result['매장수']
X = result[['주민등록인구','종사자수','사업체수']]
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeRegressor
tree = DecisionTreeRegressor()
```

```python
cross_val_score(tree,X,y,cv=8)
```

```python
cross_val_score(tree,X,y,cv=5)
```
