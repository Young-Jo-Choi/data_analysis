```python
# 사용자별로 다른 색으로 경로를 시각화하기 위해 미리 필요한 변수를 선언
Color = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'pink','gray', 'black']
users=[]
for i in range(10):
    user = 'User0'+str(i)
    users.append(user)
print(users)
```

```python
# 다음 코드를 사용해 경로를 중첩된 리스트로 만들 수 있으나 for문이 중첨되있으므로 다른 방법 사용
#bike_path=[]
#for user in users:
#    temp = df_path[path['사용자 정보']==user]
#    temp_list=[]
#    for i in temp.index:
#        latlng=[]
#        latlng.append(temp['위도'][i])
#        latlng.append(temp['경도'][i])
#        temp_list.append(latlng)
#    bike_path.append(temp_list)
```

```python
bike_path=[]
for user in users:
    # numpy 배열로 DataFrame의 값을 한번에 가져옴
    temp = df_path[df_path['사용자 정보']==user][['위도','경도']].values
    bike_path.append(temp)
```

```python
print(len(bike_path),'\n',bike_path[3])
```

```python
# folium에 한번에 시각화
map = folium.Map(location=[df_path['위도'].mean(), df_path['경도'].mean()], zoom_start=11)
for k,each_path in enumerate(bike_path):
    folium.PolyLine(locations = each_path, tooltip = 'PolyLine',color=Color[k]).add_to(map) 
for n in df_place.index:
    folium.Marker([df_place.loc[n, '대여위도'], df_place.loc[n, '대여경도']],popup=df_place.loc[n,'사용자 정보'], 
                icon=folium.Icon(icon="fa fa-telegram",color=Color[n],prefix='fa')).add_to(map)
    folium.Marker([df_place.loc[n, '반납위도'], df_place.loc[n, '반납경도']],popup=df_place.loc[n,'사용자 정보'], 
                icon=folium.Icon(icon="fa fa-telegram",color=Color[n],prefix='fa')).add_to(map)

map
```

```python
lending_place = pd.read_csv('세종특별자치시_자전거대여소_20190813.csv',encoding='euc-kr')
lending_place.head()
```

```python
temp_df = lending_place[['자전거대여소명','위도','경도']]
temp_df
```

```python
map2 = folium.Map(location=[temp_df['위도'].mean(), temp_df['경도'].mean()], zoom_start=11)

for n in range(len(temp_df)):
    lat = temp_df['위도'].iloc[n]
    lng = temp_df['경도'].iloc[n]
    name = temp_df['자전거대여소명']
    folium.Marker([lat,lng],popup=name).add_to(map2)
```
