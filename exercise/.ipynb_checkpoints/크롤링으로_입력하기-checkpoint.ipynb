```python
# 해당 url은 위키백과, selenium을 이용해 검색하는 기능까지 구현해 볼 것임
url = 'https://ko.wikipedia.org/wiki/%EC%9C%84%ED%82%A4%EB%B0%B1%EA%B3%BC'
```

```python
from selenium import webdriver
driver = webdriver.Chrome('C:/Users/hp/Desktop/최영조/chromedriver.exe')
driver.get(url)
assert "facebook" in driver.title       # 해당 사이트가 wikipedia가 아니면 에러메세지를 내라는 의미(좀 더 보완 필요)
```

```python
# search는 위키백과 입력창의 input의 이름
search_box = driver.find_element_by_name("search")  #find_element_by~는 커서를 그곳에 두는 것을 의미
                                                    #개발자 도구에는 해당하는 name, xpath등을 복사하는 기능이 있다.
search_box.send_keys("제 2차 세계대전")             # 해당 파라미터를 커서가 위치하는 곳에 입력
search_box.submit()
```

```python
html = driver.page_source
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(html,'html.parser')
```

```python
soup.get_text() == soup.text
```
