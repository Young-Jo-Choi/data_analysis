```python
import random
balance = 5000
for i in range(1000):
    a = random.randint(1,65)
    if a==1:
        balance += 24
    elif a<=10:
        balance += 3
    elif a<=37:
        balance += 0
    else:
        balance -= 1
print(balance)
```

```python
import time
start = time.time()

balances=[]
for i in range(500):
    balance = 5000
    for i in range(1000):
        a = random.randint(1,65)
        if a==1:
            balance += 24
        elif a<=10:
            balance += 3
        elif a<=37:
            balance += 0
        else:
            balance -= 1
    balances.append(balance)
    
print(sum(balances)/500)
print(time.time()-start)
```

```python
start = time.time()
A = np.random.randint(1,65,size=(500,1000))
B = np.ones_like(A)
B[A==1] = 24
B[(A>1) & (A<=10)] = 3
B[(A>11) & (A<=37)] = 0
B[A>37] = -1
C = B.cumsum(axis=1)
print(np.average(C[:,-1]+5000))
print(time.time()-start)
```
