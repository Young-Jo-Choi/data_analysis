## 멜론차트 상위에 랭크되어 있는 곡들의 정보를 가져오기(2020.04.02)

```python
# 멜론 차트 페이지
url = 'https://www.melon.com/chart/index.htm'
```

```python
from selenium import webdriver
from bs4 import BeautifulSoup
```

```python
# chromedriver의 위치 주소
driver = webdriver.Chrome('C:/Users/hp/Desktop/최영조/chromedriver.exe')

#해당 페이지를 정보 가져오기 위한 형태로 열기
driver.get(url)
```

```python
# 해당 페이지의 html 소스를 문자열 형태로 가져옴
html = driver.page_source
html[:1000]
```

```python
# html 코드를 원하는 형태로 쪼개기(?)
soup = BeautifulSoup(html,'html.parser')
```

## 현재 1위 곡은 '엠씨더맥스'의 '처음처럼', 2위 곡은 '가호'의 '시작'
### html 코드를 한번에 파악하기는 어렵기 때문에 차근차근 범위를 좁혀나갈 것임

```python
soup.select('div.wrap_song_info')
```

```python
len(soup.select('div.wrap_song_info'))
```

```python
soup.select('div.wrap_song_info>div>span>a')
```

현재 결과가 list로 출력된다. 0포함 짝수 번에 곡에 대한 정보,1부터 홀수 번에 가수에 대한 정보가 담겨있음<br>
0번과 1번은 greatness라는 이벤트로 올라간 것이니 생략하도록 하자

```python
tags = soup.select('div.wrap_song_info>div>span>a')[2:]
tags
```

```python
# 각 원소들은 tag인데 그 중 title 속성만을 출력
tags[0]['title']
```

```python
# 재생이라는 부분은 빼버리도록 하자.
tags[0]['title'][:-3]
```

```python
# 가수에 대해서도 마찬가지
tags[1]['title']
```

```python
tags[1]['title'][:-9]
```

```python
len(tags)
```

tags가 206개인데 이는 103순위까지의 노래를 담고 있음을 알 수 있음

```python
songs = []
singers = []
for i in range(len(tags)):
    if i%2 == 0:
        song = tags[i]['title'][:-3]
        songs.append(song)
    else:
        singer = tags[i]['title'][:-9]
        singers.append(singer)
```

```python
if len(songs)==len(singers):
    print("곡들의 수: ",len(songs))
else:
    print("오류가 있으니 다시 해주세요.")
```

```python
from datetime import datetime
```

```python
now = datetime.now()
```

```python
print("{}년 {}월 {}일 기준 멜론차트 순위".format(now.year,now.month,now.day))
for i,(song,singer) in enumerate(zip(songs,singers)):
    print('{}위 곡:{} written by {}'.format(i+1,song,singer))
```

### 해당 내용을 DataFrame으로 바꾸고 excel 파일로 저장하기

```python
import pandas as pd
```

```python
df = pd.DataFrame({'song':songs,'singer':singers})
df
```
