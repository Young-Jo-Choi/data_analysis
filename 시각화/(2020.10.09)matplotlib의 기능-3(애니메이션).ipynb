애니메이션을 그릴 매 프레임 마다 호출될 갱신(update) 함수를 정의
<br>FuncAnimation 인스턴스를 생성하여 그래프에 애니메이션을 추가<br>

FuncAnimation 생성자는 figure, 갱신 함수, 그 외의 파라미터를 수용한다. <br>
다음 코드에서는 각 프레임간 20ms의 시간차가 있는 100개의 프레임으로 구성된 애니메이션에 대한 인스턴스를 만든다.<br>
애니메이션의 각 프레임마다 FuncAnimation 는 갱신 함수를 호출하고, 프레임 번호를 num (이 예에서는 0~99의 범위) 으로서 전달된다.<br>
또한 갱신 함수의 추가적인 두 파라미터는 FuncAnimation 생성시 fargs에 넣어준 값이 된다.

작성한 갱신 함수는 선을 구성하는 데이터를 0 ~ num 데이터로 설정한다. (따라서 데이터가 점진적으로 그려진다).<br>

```python
# 다음 그림을 애니메이션으로 재구성

x = np.linspace(-1, 1, 100)
y = np.sin(x**2*25)
data = np.array([x, y])

plt.axis([-1.1, 1.1, -1.1, 1.1]) # xlim, ylim 비슷한 역할
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)

plt.plot(x,y,'r')
```

```python
data.shape
```

```python
# 데이터는 이미 있는 x,y,data를 재활용

fig = plt.figure()
plt.axis([-1.1, 1.1, -1.1, 1.1])
plt.plot([-0.5, 0.5], [0, 0], "b-", [0, 0], [-0.5, 0.5], "b-", 0, 0, "ro")
plt.grid(True)
plt.title("Marvelous animation")
# 여기까지는 애니메이션의 기본 상태가 되는 그림

line, = plt.plot([], [], "r-") # 비어있는 plot으로 시작한다.

# 매 갱신마다 호출될 함수
def update_line(num, data, line):
    line.set_data(data[:, :num])  # we only plot the first `num` data points.
    return line,

# 프레임 번호(0~99)가 num으로 전달
line_ani = animation.FuncAnimation(fig, update_line, frames=100, fargs=(data, line), interval=67)
plt.close() # close하지 않으면 애니메이션 하나, 기본 그림 하나가 그려질 것
line_ani
```

# 애니메이션을 비디오로 저장

비디오로 저장하기 위해서 Matplotlib은 써드파티 라이브러리(FFMPEG 또는 ImageMagick)에 의존한다. <br>
둘 중 하나라도 설치된 상태에서 animation을 만들어 저장할 수 있다. FFMPEG는 mp4, ImageMagick은 GIF 형식으로 저장이 가능하다.<br>
(코랩으로 실행 가능)

```python
# 해당 라이브러리가 설치되어있지 않으므로 코드만 첨부
Writer = animation.writers['ffmpeg']
writer = Writer(fps=15, metadata=dict(artist='Me'), bitrate=1800)
line_ani.save('my_wiggly_animation.mp4', writer=writer)
```
