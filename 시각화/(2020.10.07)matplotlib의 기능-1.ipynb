# 2. Font
폰트 설정 : plt.suptitle, ax.title, ax.xlabel, axes.xticklabels, ax.annotate, ax.text 등
## 2.1. mpl.font_manager.FontProperties
rc,rcParams 등은 전역적인 기본값 설정이지만 다음 방법으로 부분적으로 폰트 설정이 가능하다.

```python
# font setting
font_setting0 = mpl.font_manager.FontProperties()
font_setting0.set_family('Dejavu Sans') # 'serif', 'sans-serif', 'cursive', 'fantasy', 'monospace'
font_setting0.set_size(20)
font_setting0.set_style('normal') # 'normal', 'oblique', 'italic'
font_setting0.set_weight('bold')

fig, ax = plt.subplots(figsize=(5,3))

data = [10, 24, 30, 50, 40]
ax.plot(data)
ax.set_title("title", fontproperties=font_setting0)

plt.show()
```

```python
fig, ax = plt.subplots(figsize=(5,3))

data = [10, 24, 30, 50, 40]
ax.plot(data)
ax.set_title("title", fontproperties=font_setting0)
ax.set_xlabel('fff',fontproperties=font_setting0)

plt.show()
```

## 2.2. fontdict

```python
# font setting
fontdict = {
    'family': 'serif',
    'size': 15,
    'backgroundcolor': 'yellow',
    'color': 'blue',
    'weight': 'normal',
    'verticalalignment': 'baseline',
    'horizontalalignment': 'left'
}

fig, ax = plt.subplots(figsize=(5,3))

data = [10, 24, 30, 50, 40]
ax.plot(data)
ax.set_title("title", fontdict=fontdict)

plt.suptitle("suptitle", fontdict=fontdict)
plt.subplots_adjust(top=0.7)
plt.show()
```

# 3. Color

```python
c1 = 'darkred'             # matplotlib color name
c2 = '#AABBFF'             # hex code without alpha 
c3 = '#AABBFF55'           # hex code with alpha
c4 = [0.2, 0.4, 0.5]       # RGB
c5 = [0.2, 0.4, 0.5, 0.6]  # RGBA: RGB + alpha
```

```python
import pandas as pd
import numpy as np

colors = pd.DataFrame(columns=['input', 'to_hex(F)', 'to_hex(T)', 
                               'to_rgb()', 'to_rgba(alpha=None)', 'to_rgba(alpha=0.2)'])

for i in range(5):
    # to_hex()
    hex_code0 = mpl.colors.to_hex(eval(f'c{i+1}'), keep_alpha=False)
    hex_code1 = mpl.colors.to_hex(eval(f'c{i+1}'), keep_alpha=True)
    
    # to_rgb()
    rgb_code0 = np.round(mpl.colors.to_rgb(eval(f'c{i+1}')), 3)
    
    # to_rgba()
    rgb_code1 = np.round(mpl.colors.to_rgba(eval(f'c{i+1}')), 3)
    rgb_code2 = np.round(mpl.colors.to_rgba(eval(f'c{i+1}'), 0.2), 3)
    
    colors.loc[i] = [eval(f'c{i+1}'), hex_code0, hex_code1, rgb_code0, rgb_code1, rgb_code2]
    
colors
```

to_hex()는 keep_alpha 옵션에 따라 alpha 채널이 삭제/추가되는 반면<br>
to_rgb()는 alpha 채널이 자동적으로 삭제되고<br>
to_rgba()는 alpha 옵션에 따라 alpha 값이 원래 값 vs 지정 값으로 설정된다..<br>

```python
# 대표적인 color 표현방법인 RGB 컬러와 HSV 컬러를 쉽게 변환할 수 있다.

cc = [0.2,0.4,0.5]
hsv_cc = mpl.colors.rgb_to_hsv(cc)
rgb_cc = mpl.colors.hsv_to_rgb(hsv_cc)
print('cc : {} ,hsv_cc : {}, rgb_cc : {}'.format(cc,hsv_cc,rgb_cc))
```

# 4. colormaps

```python
from sklearn.datasets import load_boston
import pandas as pd

boston = load_boston()
df = pd.DataFrame(boston.data,columns=boston.feature_names)
df['MEDV'] = boston.target
df.head()
```

[공식홈페이지](https://matplotlib.org/tutorials/colors/colormaps.html)에서 다양한 colormap 정보를 확인할 수 있다.

```python
fig,ax = plt.subplots(1,3,figsize=(12,6))
ax[0].scatter(df['DIS'],df['MEDV'],s=5)
ax[0].set_xlabel('DIS')
ax[0].set_ylabel('MEDV')
ax[0].set_title('NO COLOR ADDED')

ax[1].scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5)
ax[1].set_xlabel('DIS')
ax[1].set_ylabel('MEDV')
ax[1].set_title('color = age')

# colormap 변경
ax[2].scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5,cmap='jet')
ax[2].set_xlabel('DIS')
ax[2].set_ylabel('MEDV')
ax[2].set_title('different colormap')

```

## 4.1. colorbar

```python
fig,ax = plt.subplots(1,1)
im = ax.scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5,cmap='jet')
ax.set_xlabel('DIS')
ax.set_ylabel('MEDV')
ax.set_title('different colormap')

# im으로 지정한 axes에 대해 colorbat 생성
plt.colorbar(im)
```

colorbar를 별도의 객체로 저장하고 set_label()로 label을 지정할 수 있다.

```python
fig,ax = plt.subplots(1,1)
im = ax.scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5,cmap='jet')
ax.set_xlabel('DIS')
ax.set_ylabel('MEDV')
ax.set_title('different colormap')

cbar = plt.colorbar(im)
cbar.set_label('possible!!')
plt.tight_layout()
```

### 4.1.1. colorbar의 제어
1. axes
    - orientation
    - fraction : colorbar가 차지하는 영역의 비율
    - aspect : colorbar의 긴 변,짧은 변 비율
2. colorbar
    - ticks
    - format : tick values 형식
    - drawedges : 구간마다 윤곽선 표시 

다음과 같이 여러 개의 axes 중 어디에 colorbar를 추가할지 지정할 수 있다.

```python
fig,ax = plt.subplots(1,3,figsize=(12,4))
ax[0].scatter(df['DIS'],df['MEDV'],s=5)
ax[0].set_xlabel('DIS')
ax[0].set_ylabel('MEDV')
ax[0].set_title('NO COLOR ADDED')

im = ax[1].scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5)
ax[1].set_xlabel('DIS')
ax[1].set_ylabel('MEDV')

im2 = ax[2].scatter(df['DIS'],df['MEDV'],c=df['AGE'],s=5,alpha=0.5,cmap='jet')
ax[2].set_xlabel('DIS')
ax[2].set_ylabel('MEDV')
ax[2].set_title('different colormap')

# drawedges로 1단위로 눈금 생성 -> 그렇기 때문에 좀 어두워진다.
cbar = plt.colorbar(im, ax=ax[1], ticks=[15, 50, 60], format='%1.2e', drawedges=True)
ax[1].set_title('colorbar')

cbar = plt.colorbar(im2,ax=ax[2],orientation='horizontal',aspect=10,fraction=0.5)
cbar.set_label('possible!!')
plt.tight_layout()
```

# 5. 서로 다르게 생긴 subplot (grid를 이용)

```python
import numpy as np
x = np.linspace(-1.4, 1.4, 30)
```

```python
# 격자를 도화지에 준비(그래프를 그리기 전에 먼저 틀을 잡아둔다.)
grid = plt.GridSpec(2, 2)

ax1 = plt.subplot(grid[0, 0])
ax2 = plt.subplot(grid[0, 1])

# 2행 전체
ax3 = plt.subplot(grid[1, 0:]) 

ax1.plot(x, x)
ax2.plot(x, x**2)
ax3.plot(x, x**3)
```

## 5.1. subplot2grid
보다 복잡한 부분 그래프의 위치 선정이 필요하다면, subplot2grid를 대신 사용할 수 있다. 격자의 행과 열의 번호 및 격자에서 해당 부분 그래프를 그릴 위치를 지정해줄 수 있다. (좌측상단 = (0,0)). 몇 개의 행/열로 확장되어야 하는지도 추가적으로 지정할 수 있다.

```python
# object oriented API

# 2x2의 격자를 사용, 크기는 3x3
gridsize = (3, 3)
ax1 = plt.subplot2grid(gridsize, (0,0), rowspan=2, colspan=2)
ax2 = plt.subplot2grid(gridsize, (0,2))
ax3 = plt.subplot2grid(gridsize, (1,2), rowspan=2)
ax4 = plt.subplot2grid(gridsize, (2,0), colspan=2)

ax1.plot(x, x**2)
ax2.plot(x, x**3)
ax3.plot(x, x**4)
ax4.plot(x, x**5)

plt.tight_layout()
```

```python
# 동시에 두 개의 도화지를 표현하는 것도 가능
fig1, ax1 = plt.subplots(nrows=2, ncols=1)

ax1[0].plot(x, x**2)
ax1[0].set_title("Square and Cube")

ax1[1].plot(x, x**3)


fig2, ax2 = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
ax2[0].plot(x, x**4)
ax2[0].set_title("y = x**4")

ax2[1].plot(x, x**5)
ax2[1].set_title("y = x**5")

# 첫 번쨰 도화지로 돌아가기
ax1[1].plot(x, -x**3, "r:")

# ptplot API에서는 plt.figure(1)->plt.figure(2)->plt.figure(1)과 같은 방식으로 제어 가능
```

# 6. text, annotate

```python
x = np.linspace(-3,3,500)
y = x**2

x_point = 2.5
y_point = x_point**2

fig,ax = plt.subplots()
ax.plot(x,y,'g-',linewidth=1)
ax.scatter(x_point,y_point,marker='o',c='r',linewidths=3)

# text를 box 형식으로 넣을 수 있다.
# ha는 수평적 정렬

bbox_props = dict(boxstyle="round4,pad=1,rounding_size=0.2", ec="black", fc="#EEEEFF", lw=5)
ax.text(1, 8, "이것저것 그리는 중", bbox=bbox_props, ha="right")

bbox_props = dict(boxstyle="rarrow,pad=0.3", ec="b", lw=2, fc="lightblue")
ax.text(x_point-0.4, y_point, "text in a blue box", bbox=bbox_props, ha="right",fontstyle='italic',size=15)

# 그래프 요소에 주석 달기: 관심있는 부분을 설정하고 텍스트의 위치를 설정, arrowprops를 이용하면 화살표 그리기도 가능
arrowprops = {"facecolor": "lightgreen"}
ax.annotate('점',xy=(x_point,y_point),xytext=(2.4,3.9),arrowprops=arrowprops,weight='heavy',color='blue',size=18)
```
