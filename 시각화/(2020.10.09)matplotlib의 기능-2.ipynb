# 1. ticks and tickers

```python
x = np.linspace(-2,2,100)

fig,ax = plt.subplots(1,3,figsize=(15,4))
for axes in ax:
    axes.plot(x,x**3)
    axes.grid(True)

ax[0].set_title('Default')

ax[1].set_title("x축의 tick만 조정")
ax[1].set_xticks([-2,-1,0,1,1.5,1.7])

ax[2].set_title("y축은 라벨링까지")
ax[2].set_xticks(np.arange(-2,2,0.3))
# labeling은 ticks로 설정된 값들에 대해
ax[2].set_yticks(range(-7,7,3))
ax[2].set_yticklabels(['너무 작은값', '조금 작은값','그냥','괜찮은값','큰값'],rotation=45)

plt.tight_layout()
```

# 2. polar coordinate
부분 그래프를 생성시 projection 속성을 'polar'로 설정해 극좌표계로 투영시킬 수 있다.

```python
theta = np.linspace(0, 2*np.pi, 1000)
plt.plot(theta, np.sin(theta),'r-')
plt.plot(theta, np.cos(theta),'g--')
plt.grid('True')
plt.title('일반좌표계')
```

```python
# polar 속성을 입힘
plt.figure(figsize=(5,5))
plt.subplot(111,projection='polar')
plt.plot(theta, np.sin(theta),'r-')
plt.plot(theta, np.cos(theta),'g--')
# 그리드를 그려도 똑같음
plt.title('극좌표계')
```

# 3. 3차원 투영
3d 투영을 등록하는 Axes3D를 import<br>
projection 속성이 '3d'인 그래프를 생성 -> Axes3DSubplot 객체 생성<br>
이 객체의 plot_surface 메서드를 이용하면 x,y,z 좌표를 모두 그릴 수 있다.

다음 예시 : $z=sin(\sqrt{x^2+y^2})$

```python
from mpl_toolkits.mplot3d import Axes3D
```

```python
x = np.linspace(-5, 5, 50)
y = np.linspace(-5, 5, 50)
# x와 y가 1차원인데 z축을 계산하기 위해 차원을 확장 
X, Y = np.meshgrid(x, y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

figure = plt.figure(1, figsize = (12, 4))

subplot3d = plt.subplot(111, projection='3d')

# Z가 2차원 데이터이어야한다.
surface = subplot3d.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=mpl.cm.coolwarm, linewidth=0.1)
```

```python
print('x.shape : {}, y.shape : {}, X.shape : {}, Y.shape : {}, Z.shape : {}'.format(x.shape,y.shape,X.shape,Y.shape,Z.shape))
```

```python
# 동일한 데이터를 2차원 평면으로(contour나 pcolor를 이용)
fig,ax = plt.subplots()
ax.contour(X,Y,Z,cmap=mpl.cm.coolwarm)
axes = ax.pcolor(X,Y,Z,cmap=mpl.cm.coolwarm)

cbar = plt.colorbar(axes,ax=ax)
```

# 4. 히스토그램
bin을 조정할 수 있다.

```python
data = [1, 1.1, 1.8, 2, 2.1, 3.2, 3, 3, 3, 2.9, 3, 3.7]
fig,ax = plt.subplots(1,2,figsize=(8,4))
ax[0].hist(data,bins=10,rwidth=0.8)
ax[0].set_xlabel('value')
ax[0].set_ylabel('빈도')
ax[0].set_title('기본값')

ax[1].hist(data,bins=[1,1.5,2.3,3.1,3.7],rwidth=0.8)
ax[1].set_xlabel('value')
ax[1].set_ylabel('빈도')
ax[1].set_title('bins를 임의로 조정')
```

```python
# 각 histogram의 모양(histtype)에 따른 비교, 데이터는 전부 같은 모양의 분포

fig,ax=plt.subplots(2,2,figsize=(8,8))
data = np.random.randn(400)

ax[0,0].hist(data, bins=5, color='g', alpha=0.75, label='bar hist') 
ax[1,0].hist(data, color='b', alpha=0.65, histtype='stepfilled', label='stepfilled hist')
ax[0,1].hist(data, color='r', histtype='step', label='step hist')
ax[1,1].hist((data, data+10), color=('r','m'), alpha=0.55, histtype='barstacked', label=('barstacked a', 'barstacked b'))

for axes in ax.ravel():
    axes.legend(loc='upper right')
```
