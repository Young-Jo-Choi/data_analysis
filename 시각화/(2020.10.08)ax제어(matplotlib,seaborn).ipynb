```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
from matplotlib import font_manager,rc
import platform

if platform.system() == 'Windows':
    path = 'c:/windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font',family = font_name)
elif platform.system() == 'Darwin':
    rc('font',family = 'AppleGothic')
else:
    print('Check your OS system')
    
%matplotlib inline
```

그래프의 기본적인 font,grid style 등 설정은 plt.rcParams에 정의되어있다.<br>
- plt.style.use()를 사용하면 background color, grid color 등 설정이 편리하다.<br>
- seaborn의 seaborn.set_context()를 사용하면 fontsize, linewidth등 설정에 용이하다.

```python
plt.style.use('seaborn-whitegrid')
sns.set_context('talk')

fig, ax = plt.subplots(figsize=(7,3))
```

```python
temp_dict = {'가장동':15000,
            '괴정동':24000,
            '갈마동':12700,
            '둔산동':32000,
            '용문동':18900,
            '은행동':22670,
            '태평동':12100,
            '유천동':20400,
            '월평동':21280,
            '내동':16700,
            '변동':19720,
            '도마동':29100}
temp_name = list(temp_dict.keys())
temp_numeric = list(temp_dict.values())
```

```python
fig,ax = plt.subplots()
ax.barh(temp_name,temp_numeric)
```

```python
# dictionary를 DataFrame으로
df = pd.DataFrame.from_dict(temp_dict, orient='index', columns=['numeric'])
df
```

```python
df.plot(kind='barh')
plt.axvline(30000)
```

```python
# 타이틀 폰트에 관한 dictionary를 미리 만들어 사용하면 일일이 지정하는 것보다 편리하다.
font_title = {
    'family': 'serif',
    'size': 20,
#     'backgroundcolor': 'yellow',
    'color': 'brown',
    'weight': 'bold',
    'verticalalignment': 'baseline',
    'horizontalalignment': 'center'
}


fig,ax = plt.subplots()
sns.barplot(temp_numeric,temp_name,ax=ax)
ax.set_title('temp_title',fontdict = font_title)

# set_xticklabels는 축의 라벨을 조정, 단 labels에 새로운 이터러블 객체를 입력해야함
# ax.get_xticks()를 쓰면 원래 x축의 라벨들을 추출할 수 있음

ax.set_xticklabels(labels = ax.get_xticks().astype('int'), rotation=45)
```

```python
# ax.get_~ 하면 추출할 수 있는게 많다.
print(ax.get_xticks())
print(ax.get_xlim())
```

```python
fig,ax = plt.subplots()
sns.barplot(temp_numeric,temp_name,ax=ax)
ax.set_title('temp_title',fontdict = font_title)

# 웬만하면 xtick의 이름이 바뀌기 전에 제한을 걸도록 하자
ax.set_xlim(10000,35000)

temp_mean = np.mean(temp_numeric)
ax.axvline(temp_mean,ls='-', lw=3, color='red')

arrowprops = {'arrowstyle': '->'}
ax.annotate('평균',xy=(temp_mean,5),xytext=(temp_mean+2500,5),fontsize=15,arrowprops=arrowprops)

# bar에 글자 표시해보기
for i in [3,5,11]:
    ax.text(x=12000,y=(i+0.2),s='놀기 좋은 곳',fontstyle='italic')
    
# list comprehension을 통해 xtick을 조금 변경
ax.set_xticklabels([f'{xlabel/1000:0.0f}천' for xlabel in ax.get_xticks()])

```
