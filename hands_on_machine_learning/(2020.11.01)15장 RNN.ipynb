# 순환 뉴런과 순환 층
기본적으로 피드포워드 신경망과 매우 비슷하지만 뒤쪽으로 순환하는 연결이 존재한다는 점이 다르다.<br>
x : 입력<br>
y : 출력<br>
하나의 층에서 하는 계산은 다음과 같다.<br>
$\cdot\cdot\cdot\cdot$<br>
$f_{t-2}(x(t-2),y(t-3)) = y(t-2)$<br>
$f_{t-1}(x(t-1),y(t-2)) = y(t-1)$<br>
$f_{t}(x(t),y(t-1)) = y(t)$<br>

각 타임스텝 t(혹은 프레임)마다 x(t)와 이전 타임 스탭의 출력인 y(t-1)을 입력으로 받는다. 첫 번째 타임 스텝에서는 이전 출력이 없으므로 일반적으로 0으로 설정한다.<br>이런 방식을 시간에 따라 네트워크를 펼쳤다(unrolling the network through time)고 표현한다.

각 순환 뉴런은 두 벌의 가중치를 가진다. 하나는 입력 x(t)를 위한 것이고 다른 하나는 이전 타임 스텝의 출력 y(t-1)을 위한 것이다. 이 가중치 벡터를 $w_x$와 $w_y$라고 하겠다. 하나의 순환 뉴런이 아니라 순환 층 전체를 생각하면 가중치 벡터를 가중치 행렬 $W_x$와 $W_y$로 바꿀 수 있다.<br>$\phi$는 ReLU와 같은 활성화 함수

$y_{(t)} = \phi({W_x}^Tx_{(t)} + {W_y}^Ty_{(t-1)} + b)$

$Y_{(t)} = \phi(X_{(t)}W_x + Y_{(t-1)}W_y + b)$

피드포워드 신경망처럼 타임 스텝 t에서도 모든 입력을 행렬 $X_{(t)}$로 만들어 미니배치 전체에 대해 순환 층의 출려값을 한 번에 계산할 수 있다.<br>
(m은 미니배치의 샘플의 수)
- $X_{t}$는 모든 샘플의 입력값을 담은 $m$ x $n_{inputs}$ 행렬이다.
- $Y_t$는 타임 스텝 t에서 미니배치에 있는 각 샘플에 대한 층의 출력을 담은 $m$ x $n_{neurons}$ 행렬이다.
- $W_x$는 현재 타임 스텝의 입력에 대한 연결 가중치를 담은 $n_{inputs}$ x $n_{neurons}$ 행렬이다.
- $W_y$는 이전 타임 스텝의 입력에 대한 연결 가중치를 담은 $n_{inputs}$ x $n_{neurons}$ 행렬이다.

타입 스텝에 걸쳐서 어떤 상태를 보존하는 신경망의 구성 요소를 메모리 셀(memory cell)이라고 부른다. 셀의 상태는 $h_t = f(h_{t-1}, x_t)$로 나타낸다.

## 입력과 출력 시퀀스
- "sequence to sequence" network
- "sequence to vector" network : 입력 시퀀스를 네트워크에 주입하고, 마지막을 제외한 모든 출력을 무시할 수 있다.
  - 예)영화 리뷰의 연속된 단어를 입력하면 네트워크는 감성 점수를 출력하지만 최종 출력층만 사용된다.
- "vector to sequence" network : 반대로 입력을 하나의 벡터만 주입하고, 시퀀스를 출력할 수 있다.
  - 예) 이미지(CNN 출력)를 입력하여 이미지에 대한 캡션을 출력할 수 있다.
- encoder(sequence to vector network) 뒤에 decoder(vector to sequence network)를 연결할 수 있다. 
  - 예) 번역 :  한 언어의 문장을 네트워크에 주입하면 인코더는 이 문장을 하나의 벡터 표현으로 변환하고, 그다음에 디코더가 이 벡터를 다른 언어의 문장으로 디코딩

# RNN 훈련
BPTT(backpropagation through time) : 타임 스텝으로 네트워크를 펼치고 보통의 역전파를 사용한다.<br>
비용함수 : $C(Y_{(0)},Y_{(1)}, \cdot\cdot\cdot , Y_{(T)})$ 네트워크에 따라 일부 출력을 무시할 수 있다.

```python
# 시계열 생성
def generate_time_series(batch_size, n_steps):
  freq1, freq2, offsets1, offsets2 = np.random.rand(4,batch_size,1)
  time = np.linspace(0,1,n_steps)
  series = 0.5 * np.sin((time-offsets1)*(freq1*10+10))     # 사인 곡선 1
  series += 0.2 * np.sin((time-offsets2)*(freq2*20+20))    # 사인 곡선 2
  series += 0.1*(np.random.rand(batch_size,n_steps)-0.5)   # 잡음
  return series[...,np.newaxis].astype(np.float32)
```

```python
n_steps = 50
series = generate_time_series(10000,n_steps+1)
X_train,y_train = series[:7000,:n_steps],series[:7000,-1]
X_valid,y_valid = series[7000:9000,:n_steps],series[7000:9000,-1]
X_test,y_test = series[9000:,:n_steps],series[9000:,-1]
print(X_train.shape,y_train.shape,'\n',X_valid.shape,y_valid.shape,'\n',X_test.shape,y_test.shape)
```

```python
# 평균 제곱오차
# y_pred는 X_valid의 마지막 값

y_pred = X_valid[:,-1]
np.mean(keras.losses.mean_squared_error(y_valid,y_pred))
```

또 다른 간단한 방법은 완전 연결 네트워크를 사용하는 것이다. 입력마다 1차원 배열을 기대하므로 Flatten층을 추가해야한다.

```python
model = keras.models.Sequential()
model.add(keras.layers.Flatten(input_shape=[50,1]))
model.add(keras.layers.Dense(1))

model.compile(loss='mse',optimizer='adam')

history = model.fit(X_train, y_train, epochs=20,
                    validation_data=(X_valid, y_valid))
```

동작 : 초기 상태 $h_{(init)}$을 0으로 설정하고 첫 번째 타임 스텝 $x_{0}$와 함께 하나의 순환 뉴런으로 전달한다. 뉴런은 이 값의 가중치 합을 계산하고 하이퍼볼릭 탄젠트 활성화 함수를 적용해 결과를 만들어 첫 번째 $y_0$을 출력한다. 기본 RNN에서는 이 출력이 새로운 상태 $h_o$이 된다. 이 새로운 상태는 다음 입력 값 $x_1$과 함께 동일한 순환 뉴런으로 전달된다. 이 과정이 마지막 타임 스텝까지 반복된다. <br>

그다음 이 층은 마지막 값 $y_{49}$을 출력한다. 모든 시계열에 대해 동일한 과정이 수행된다.

기본적으로 케라스의 순환 층은 최종 출력만 반환하는데 타임 스텝마다 출력을 반환하려면 return_sequence=True로 지정하면 된다.

이 모델은 순진한 예측보다는 낫지만 간단한 선형 모델을 앞지르지는 못한다. 선형모델은 각 뉴런에 대해 입력(타임 스탭)마다 하나의 파라미터를 가지고 편향 역시 가진다.(51개의 파라미터)<br>
 반면 SimpleRNN의 순환 뉴런은 입력과 은닉 상태 차원(SimpleRNN에서는 층의 순환 뉴런 개수와 같다)마다 하나의 파라미터를 가지고 편향을 갖기 때문에 총 3개의 파라미터를 갖는다.

## 심층 RNN
- 심층 RNN(deep RNN, p608 그림 참조) : 셀을 여러 층으로 쌓는다.

케라스로 구현하기 위해서는 SimpleRNN층을 여러 개 쌓으면 된다. LSTM이나 GRU 같은 다른 종류의 층을 추가할 수도 있다.

```python
# 모든 층에 return_sequence=True를 설정해야한다(마지막 출력이 관심 대상이 아니면 마지막 층에서는 층을 설정하지 않는다.). 
# 그렇지 않으면 모든 타임 스텝에 대한 출력을 담은 3D 배열이 아니라 마지막 타임 스텝의 출력만 담은 2D 배열이 출력되고 다음 순환 층이 3D 형태의 시퀀스를 받지 못한다.
# 처음 입력하는 unit 파라미터는 출력의 개수
model = keras.models.Sequential()
model.add(keras.layers.SimpleRNN(20,return_sequences=True,input_shape=[None,1]))
model.add(keras.layers.SimpleRNN(20,return_sequences=True))
model.add(keras.layers.SimpleRNN(1))

model.compile(loss='mse',optimizer='adam')
```

```python
history = model.fit(X_train,y_train,epochs=20,validation_data=(X_valid,y_valid))
```

마지막 스텝에서만 그 다음 값 10개를 예측하도록 훈련하는 대신 모든 타임 스텝에서 다음 값 10개를 예측하도록 모델을 훈련시킬 수 있다.<br>
ie) sequence-to-vector RNN을 sequence-to-sequence RNN으로 바꿀 수 있다.<br>
이 방식에서 장점은 마지막 타임 스텝에서의 출력뿐만 아니라 모든 타임 스텝에서 RNN 출력에 대한 항이 loss 함수값에 포함된다는 것이다. 곧 더 많은 오차 그래디언트가 모델로 흐른다큰 뜻이고 시간에 따라서만 흐를 필요가 없다는 것이다. 각 타임 스텝의 출력에서 그래디언트가 흐르기 때문에 훈련을 안정적이면서도 빠른 속도로 수행할 수 있게 만든다.

구체적으로는 타임 스텝 0에서 타임 스텝 1\~10의 예측을 담은 벡터를 출력할 것이고 타임 스텝 1에서는 타임 스텝 2\~11의 예측을 담은 벡터를 출력할 것이다. 각 타깃은 입력 시퀀스와 동일한 길이의 sequnce이고 이 sequence는 타임 스텝마다 10차원의 벡터를 담고 있다.

```python
Y = np.empty((10000,n_steps,10)) # 각 타깃은 10D 벡터인 시퀀스이다.
for step_ahead in range(1,10+1):
  Y[:,:,step_ahead-1] = series[:,step_ahead:step_ahead+n_steps,0]
Y_train = Y[:7000]
Y_valid = Y[7000:9000]
Y_test = Y[9000:]
```

인과 모델(causal model) : 타임 스텝마다 모델은 이전 타임 스텝만 볼 수 있고 앞을 볼 수는 없다.<br><br><br>
이 모델을 sequence-to-sequence 모델로 바꾸기 위해서는 마지막 층을 포함한 모든 층에 return_sequences=True를 지정해야 한다. 그다음 모든 타임 스텝에서 출력을 Dense 층을 적용해야 한다. 이런 복잡한 과정을 TimeDistributed 층을 통해 해결할 수 있다.<br>
TimeDistributed 층은 Dense층과 같은 다른 층을 감싸서 입력 시퀀스의 모든 타임 스텝에 적용시킨다. 

참고) 각 타임 스텝을 별개의 샘플처럼 다루도록 입력의 크기를 바꿔 효과적으로 수행하는데 <br>\[배치 크기,타임 스텝 수, 입력 차원\]에서 \[배치크기x타임 스텝 수, 입력 차원\]의 크기로 바꾼다.<br> 그다음 Dense 층을 적용 후 마지막으로 출력 크기를 \[배치크기x타임 스텝 수, 입력 차원\]에서  \[배치 크기,타임 스텝 수, 입력 차원\]로 되돌린다.

```python
model = keras.models.Sequential()
model.add(keras.layers.SimpleRNN(20,return_sequences=True,input_shape=[None,1]))
model.add(keras.layers.SimpleRNN(20,return_sequences=True))
model.add(keras.layers.TimeDistributed(keras.layers.Dense(10)))

def last_time_step_mse(Y_true,Y_pred):
  return keras.metrics.mean_squared_error(Y_true[:,-1],Y_pred[:,-1])

optimizer = keras.optimizers.Adam(lr=0.01)
model.compile(loss='mse',optimizer=optimizer,metrics=[last_time_step_mse])
```

```python
model.fit(X_train,Y_train,epochs=20,validation_data=(X_train,Y_train))
```

```python
print("모든 층의 개수 : ",len(model.layers))
for j in range(len(model.layers)):
  print("{}번째 층의 파라미터 행렬의 개수 : {}".format(j,len(model.layers[j].get_weights())))
  print("{}번째 층의 input shape : ".format(j),model.layers[j].input.shape)
  for i in range(len(model.layers[j].get_weights())):
    print("{}번째 층의 {}번째 파라미터들의 행렬 모양 :  {}".format(j,i,model.layers[j].get_weights()[i].shape))
  print("{}번째 층의 output shape : ".format(j),model.layers[j].output.shape,'\n')
```

```python
model.evaluate(X_test,Y_test)
```

Dense 층이 실제로 시퀀스를 입력으로 받을 수 있는데 마치 TimeDistributed(Dense(...))처럼 입력을 다룬다. 즉 모든 타임 스텝에 걸쳐 독립적으로 마지막 입력 차원에만 적용된다. 따라서 마지막 입력 층을 Dense(10)으로 바꿀 수 있으나 Dense 층을 타임 스텝마다 독립적으로 적용하고 모델이 하나의 벡터가 아니라 시퀀스를 출력한다는 것을 드러내기 위해 TimeDistributed(Dense(10))을 사용하도록 한다.<br>
