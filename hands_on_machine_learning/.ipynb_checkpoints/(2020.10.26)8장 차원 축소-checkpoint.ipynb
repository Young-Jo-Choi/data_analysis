고차원(예를 들면 10,000차원)의 hypercube의 경우 대다수의 점은 경계와 매우 가까이 위치한다.<br>
또한 임의의 두 점을 선택했을 때 차원이 커질수록 두 점의 거리의 평균 역시 매우 커지게 된다.(아래 계산 참조)

```python
def mean_of_two_point(dim):
    X = np.random.rand(1000,dim*2)
    square_dist = 0
    for i in range(dim):
        square_dist += (X[:,i*2]-X[:,i*2+1])**2
    dist = np.sqrt(square_dist)
    return np.mean(dist)
```

```python
mean_of_two_point(2)
```

```python
mean_of_two_point(10000)
```

이로 인해 고차원 데이터셋은 희박할 가능성이 높다. 이런 경우 예측을 위해 훨씬 많은 extrapolation을 해야하기 때문에 저차원일 때보다 예측이 더 불안정해진다. 간단히 말해 훈련 세트의 차원이 클수록 과대적합의 위험이 커진다.(차원의 저주)

이론적으로 차원의 저주를 해결하는 방법은 훈련 샘플의 밀도를 충분히 높일 수 있게 훈련 세트의 크기를 키우는 것이지만 차원이 커질수록 밀도를 채우기 위한 훈련 샘플의 수 역시 기하급수적으로 늘어난다.(간단한 MNIST 데이터만 해도 28*28차원이다.)<br>
단 100개의 차원만 고려해도 샘플들이 평균 0.1이내의 거리에 위치시키이 위해서는 $10^80$개가 넘는 샘플이 필요하다.

## 매니폴드 학습
2D 매니폴드는 고차원 공간에서 휘어지거나 뒤틀린 2D 모양이다. d차원 매니폴드는 국부적으로 d차원 초평면으로 보일 수 있는 n차원 공간의 일부이다.(d<n). 스위스 롤의 경우 d=2, n=3이다.

많은 축소 알고리즘이 훈렩 샘플이 놓여 있는 매니폴드를 모델링하는 식으로 작동한다. 이를 "매니폴드 학습(manifold learning)"이라고 한다. 이는 대부분 실제 고차원 데이터셋이 저차원 매니폴드에 가깝게 놓여있다는 가정(manifold assumption, manifold hypothesis)에 근거한다.(이미지 생성의 제약) 이런 가정은 처리해야할 작업이 저차원 매니폴드에 표현된다면 더 간단해질 것이라는 가정이 같이 병행된다. 단 저차원 투영이 반드시 결정경계를 더 간단하게 하는 것은 아니다.

매니폴드 가정이 성립할 때 차원 축소 알고리즘이 어떻게 동작하는지 살펴본다.
# PCA
PCA(주성분 분석,principal component analysis)는 먼저 데이터에 가까운 초평면을 정의한 다음, 데이터를 이 초평면에 투영시킨다.

## 분산 보존
분산이 최대한 보존되는 축을 선택하는 것이 정보가 가장 적게 손실된다.

```python
x = np.linspace(0,5,1000)
y = 3*x+2
noise = np.random.randn(1000)/4
fig,ax = plt.subplots(1,2)
ax[0].scatter(x+noise,y,s=1,c='red')
ax[0].plot(x,y,linewidth=5)
xlims = ax[0].get_xlim()
ylims = ax[0].get_ylim()
ax[1].scatter(x,noise,s=1,c='red')
ax[1].set_xlim(xlims[0],xlims[1])
ax[1].set_ylim(ylims[0]-np.mean(ylims),ylims[1]-np.mean(ylims))
```

분산이 최대로 보존되는 축을 선택하는 것이 정보가 가장 적게 손실되므로 합리적이다. 다른 방식으로 설명하면 원본 데이터셋과 투영된 것 사이의 평균 제곱 거리를 최소화하는 축이다.

## 주성분
PCA는 훈련 세트에서 분산이 최대인 축을 찾는다. 그 후에 첫 번째 축에 직교하고 남은 분산을 최대한 보존하는 두 번째 축을 찾는다. 이런 방식으로 사용자가 지정한 n번 째 축까지를 찾으며 i번째 축을 i번째 주성분(PC,principal component)라고 부른다.

수학적으로 특이값 분해(SVD, singular value decomposition)이라는 표준 행렬 분해 기술이 있어서 훈련 세트 행렬 X를 $U\sum{V^{T}}$로 분해할 수 있다.

참고) 공분산 행렬은 고유벡터의 분산이 어느 방향으로 가장 큰지 나타낸다. 중앙에 맞춰진 훈련 세트의 공분산 행렬은 <br>
$Cov = {1\over n-1}X^TX = {1 \over n-1}{(U\sum{V^{T}})}^T(U\sum{V^{T}}) = V{{\sum}^2 \over n-1} V^T$ 이다.<br>
그러므로 V가 공분산 행렬의 고유 벡터이고 주성분이 된다. 사이킷런의 PCA는 scipy.linalg.svd 함수를 사용하여 주성분 V를 구한다. 다만 scipy나 numpy에서 반환되는 U,s,V 중 V가 SVD 공식 $U\sum{V^{T}}$의 $V^T$이므로 전치 과정을 거쳐야한다.

```python
# 넘파이 코드, PCA는 데이터셋의 평균이 0이라고 가정하므로 맞춰주도록 한다. 사이킷런에서는 이 작업이 포함되어 있다.

X_centered = X - X.mean(aixs=0)
U,s,Vt = np.linalg.svd(X_centered)
c1 = Vt.T[:,0]
c2 = Vt.T[:,1]

# c1, c2는 두 개의 주성분
```

## d차원으로 투영하기
주성분을 모두 추출해 냈다면 처음 d개의 주성분으로 정의한 초평면에 투영하여 데이터셋의 차원을 d차원으로 축소시킬 수 있다.<br>
훈련 세트를 d차원으로 투영하기 : $X_{d-proj} = XW_d$,  여기서 $W_d$는 d개의 열을 가지는 행렬, 곧 d 주성분까지를 포함하는 행렬이다.

```python
W2 = Vt.T[:,:2]
X2D = X_centered.dot(W2)
```

```python
# 사이킷런
from sklearn.decomposition import PCA
pca = PCA(n_components = 2)
X2D = pca.fit_transform(X)
```

pca.components_ 속성에 $W^d$의 전치가 담겨있다.<br>

## 점진적 PCA
PCA의 문제점은 SVD 알고리즘 실행을 위해 전체 훈련 세트를 메모리에 올려야 한다는 것이다. <br>
그런 단점을 보완하기 위해 "점진적 PCA(IPCA, incremental PCA)" 알고리즘이 개발되었다. 훈련 세트를 미니배치로 나눈 뒤 IPCA 알고리즘을 한 번에 하나씩 주입한다. 이런 방식은 훈련 세트가 클 때 유용하고 온라인(실시간)으로 PCA를 적용할 수도 있다.<br>
(혹은 np.memmap을 이용해 매우 큰 배열을 메모리에 들어있는 것처럼 다루는 방식도 있다. p288 참고)

```python
from tqdm import tqdm_notebook
```

```python
# 넘파이의 array_split() 함수를 사용해 100개의 미니배치로 나누고 사이킷런의 IncreamentalPCA 클래스에 주입해 실행

from sklearn.decomposition import IncrementalPCA

n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in tqdm_notebook(np.array_split(data,n_batches)):
    inc_pca.partial_fit(X_batch)
    
X_reduced = inc_pca.transform(data)
print(X_reduced.shape)
```

# 커널 PCA
서포트 벡터 머신의 커널 트릭의 아이디어를 차용한다. 고차원 특성 공간에서 선형 결정 경계는 원본 공간에서는 복잡한 비선형 결정 경계에 해당한다. 같은 기법을 PCA에 적용해 차원 축소를 위한 복잡한 비선형 투형을 수행할 수 있다. 이를 커널PCA(kPCA,kernel PCA)라고 한다. 이 기법은 투영된 후에 샘플의 군집을 유지하거나 꼬인 매니폴드에 가까운 데이터셋을 펼칠 때도 유용하다. 

```python
# 스위스 롤에 적용
from sklearn.datasets import make_swiss_roll
roll = make_swiss_roll(n_samples=1500)
X = roll[0]
y = roll[1]
```

```python
# rbf 커널에 관해서는 p213 참고

from sklearn.decomposition import KernelPCA
lin_pca = KernelPCA(n_components = 2, kernel="linear", fit_inverse_transform=True)
rbf_pca = KernelPCA(n_components = 2, kernel="rbf", gamma=0.0433, fit_inverse_transform=True)
sig_pca = KernelPCA(n_components = 2, kernel="sigmoid", gamma=0.001, coef0=1, fit_inverse_transform=True)
```

```python
fig,ax = plt.subplots(1,3,figsize=(15,5))
for axes,kpca,name in zip(ax,[lin_pca,rbf_pca,sig_pca],['linear','rbf kernel','sigmoid kernel']):
    X_pca = kpca.fit_transform(X)
    axes.scatter(X_pca[:,0],X_pca[:,1],c=y)
    axes.set_title('{}'.format(name))
plt.tight_layout()
```

## 커널 선택과 하이퍼파라미터 튜닝
kPCA는 비지도 학습이므로 좋은 커널과 하이퍼파라미터에 대한 명확한 기준이 없다. 하지만 차원 축소는 지도 학습의 전처리 단계에서 활용되므로 그리드 탐색을 하면 주어진 문제에서 가장 성능이 좋은 커널과 하이퍼파라미터를 선택할 수 있다. 다음 파이프라인을 통해 살펴본다.

```python
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline

clf = Pipeline([
    ('kpca',KernelPCA(n_components=2)),
    ('log_reg',LogisticRegression())
])

param_grid = [{
    'kpca__gamma':np.linspace(0.03,0.05,10),
    'kpca__kernel':['rbf','sigmoid']
}]
```

```python
y_bin = y>9.5

grid_search = GridSearchCV(clf,param_grid,cv=3)
grid_search.fit(X,y_bin)
print(grid_search.best_params_)
```

단지 선형 PCA만큼 원본으로의 재구성이 쉽지 않다. 축소된 공간에 있는 샘플에 대해 선형 PCA를 역전시키면 재구성된 데이터 포인트는 원본 공간이 아닌 특성 공간에 놓이게 된다. 이 특성공간은 무한 차원이기 때문에 재구성된 포인트를 계산할 수 없고 재구성에 따른 실제 에러를 계산할 수 없다.<br>
재구성된 포인트에 가깝게 매핑된 원본 공간의 포인트는 찾을 수 있는데 이를 재구성 원상(역상)(pre-image)이라고 부른다. 원상을 얻게 되면 원본 샘플과의 제곱 거리를 측정할 수 있는데 이 오차를 최소화하는 커널과 하이퍼파라미터를 선택할 수 있다.

```python
# 재구성은 다음과 같이 할 수 있다.
# fit_inverse_transform을 True로 지정해야만 inverse_transform() 메서드를 사용할 수 있다.
rbf_pca = KernelPCA(n_components=2,kernel='rbf',gamma=0.05,fit_inverse_transform=True)
X_reduced = rbf_pca.fit_transform(X)
X_preimage = rbf_pca.inverse_transform(X_reduced)
```

# LLE
지역 선형 임베딩(LLE,locally linear embedding)은 비선형 차원 축소(nonlinear dimensionality reduction)의 기술 중 하나이다.<br>
이전처럼 투영에 의존하지 않는데 LLE는 훈련 샘플이 가장 가까운 이웃(c.n.,closet neighbor)에 얼마나 선형적으로 연관되어있는지 측정한다. 그런 다음 국부적인 관계가 가장 잘 보존되는 훈련 세트의 저차원 표현을 찾는다.<br>
이 방법은 잡음이 너무 많지 않은 경우 꼬인 매니폴드를 펼치는데 잘 동작한다.

```python
from sklearn.manifold import LocallyLinearEmbedding
lle = LocallyLinearEmbedding(n_components=2,n_neighbors=10)
# 스위스롤을 펼친다.
X_reduced = lle.fit_transform(X)
plt.scatter(X_reduced[:,0],X_reduced[:,1],c=y,cmap=mpl.cm.coolwarm)
```

스위스롤이 완전히 펼쳐졌고 지역적으로는 샘플 간 거리가 잘 보존되어있다. 그러나 크게보면 샘플간 거리가 잘 유지되어 있지 않다. 오른쪽은 압축되어있고 왼쪽은 확장되어 있다. 그럼에도 불구하고 LLE는 매니폴드를 모델링하는데 잘 동작한다.

(수식은 생략, p294 참고)
대량의 데이터셋에서는 적용하기 어렵다.

# 다른 기법

## t-SNE
비슷한 샘플은 가까이, 비슷하지 않은 샘플은 멀리 떨어지도록 하면서 차원을 축소, 주로 시각화에 많이 사용되며 특히 고차원 공간에 있는 샘플의 군집을 시각화할 때 사용된다.
## LDA
사실 분류 알고리즘. 훈련 과정에서 클래스 사이를 가장 잘 구분하는 축을 학습한다. 이 축은 데이터가 투영되는 초평면을 정의하는데 사용할 수 있다. 이 알고리즘의 장점은 투영을 통해 가능한 한 클래스를 멀어지게 유지시키므로 SVM 분류기 같은 다른 분류 알고리즘을 적용하기 전에 차원을 축소시키는 데에 좋다.

```python
print(X_train.shape,X_test.shape,y_train.shape,y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
```

```python
start = time.time()
rf.fit(X_train.reshape(-1,28*28),y_train)
print('성능 : ',rf.score(X_test.reshape(-1,28*28),y_test), '\n걸린 시간 : ',time.time()-start)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
start = time.time()
X_reduced = pca.fit_transform(X_train.reshape(-1,28*28))
X_test_reduced = pca.transform(X_test.reshape(-1,28*28))
mid = time.time()
rf.fit(X_reduced,y_train)
end = time.time()
print('성능 : {}\nPCA까지 걸린 시간 : {}, PCA 이후 끝까지 걸린 시간 : {}'.format(rf.score(X_test_reduced,y_test),
                                                               mid-start,end-mid))
```

```python
# 반드시 속도가 감소하는 것은 아니다. 이번에는 로지스틱회귀를 사용
from sklearn.linear_model import LogisticRegression
log = LogisticRegression()
start = time.time()
log.fit(X_train.reshape(-1,28*28),y_train)
print('성능 : ',log.score(X_test.reshape(-1,28*28),y_test), '\n걸린 시간 : ',time.time()-start)
```

```python
start = time.time()
X_reduced = pca.fit_transform(X_train.reshape(-1,28*28))
X_test_reduced = pca.transform(X_test.reshape(-1,28*28))
mid = time.time()
log.fit(X_reduced,y_train)
end = time.time()
print('성능 : {}\nPCA까지 걸린 시간 : {}, PCA 이후 끝까지 걸린 시간 : {}'.format(log.score(X_test_reduced,y_test),
                                                               mid-start,end-mid))
```

랜덤포레스트일 때는 아니지만 로지스틱회귀에 대해서는 큰 속도 개선이 확인된다. 성능 역시 별로 감소하지 않는다.

```python
from sklearn.pipeline import Pipeline
pipe = Pipeline([
    ('pca',PCA(n_components=0.95)),
    ('lle',LocallyLinearEmbedding(n_components=2))
])
```

```python
X_list = []
X_flat = X.reshape(-1,28*28)
for redu in tqdm_notebook([pca,LLE,pipe]):
    start = time.time()
    X_redu = redu.fit_transform(X_flat)
    X_list.append(X_redu)
    print('걸린 시간 : ',time.time()-start)
```
