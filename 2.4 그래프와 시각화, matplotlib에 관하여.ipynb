# 서브플롯 간의 간격 조절하기

```python
# 다음은 서브플롯 간의 간격을 주지 않는 그래프
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
# ax.plot(x,y,'g--')는 색상과 선 스타일을 지정하는 방법
# ax.plot(x,y,linestyle=='--',color='g')로 좀 더 명시적인 방법으로 가능함

#아래는 그래프를 그리는 다양한 방법들
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
axes[0,0].plot(np.random.randn(30).cumsum(),'ko--')
axes[1,0].plot(np.random.randn(30).cumsum(), color='g',linestyle = 'dashed',marker='*')

data = np.random.randn(30).cumsum()
axes[0,1].plot(data, color='b',linestyle = 'dashed',marker='d',label='Default')
axes[0,1].plot(data, color='r',label='steps-post', drawstyle='steps-post')
axes[0,1].legend(loc='best')            #범례가 가장 적절한 위치에 추가됨(location을 best로 설정)
```

# 눈금, 라벨, 범례

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)     #하나짜리 그래프를 생성
ax.plot(np.random.randn(1000).cumsum())        #1000의 무작위 값들의 누적값의 점들 잇는 선 그리기

# x축의 눈금, 라벨
ticks = ax.set_xticks([0,250,500,750,1000])   #x축의 눈금이 이 5개 값으로 변경됨
labels =ax.set_xticklabels(['one','two','three','four','five'],fontsize='small',rotation=30)  #같은 위치값 찾아 축이름 변경
                                                     #fontsize                                #30도 가량 회전시킴
ax.set_title('My first matplotlib plot')   # 제목
ax.set_xlabel('Stage')      # x축의 라벨

# y축의 경우도 똑같이 진행 가능
```

```python
# axes의 set 이라는 메서드 하나로 할 수도 있음
props = {'title': 'My second','ylabel':'random number'}
ax.set(**props)
```

```python
# 범례
fig = plt.figure(); ax = fig.add_subplot(1,1,1)  
data1 = np.random.randn(1000).cumsum()
data2 = np.random.randn(1000).cumsum()
data3 = np.random.randn(1000).cumsum()
ax.plot(data1,'k',label='one')
ax.plot(data2,'k--',label='two')
ax.plot(data3,'k.',label='three')
#plt.legend() 혹은 ax.legend()
ax.legend(loc='best')            
# -->범례추가, loc는 범례의 위치
# 범례가 불필요하다면 label='_nolegend_' 옵션 사용
```

# 주석과 그림 추가하기

```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = pd.read_csv('spx.csv',index_col=0, parse_dates=True)      #컬럼 두개짜리 단순하지만 양이 많은 data
spx = data['SPX']

spx.plot(ax=ax, style='k-')         #그래프 그릴 때 y값들에 대한 메서드처럼 plot을 사용할 수도 있다

crisis_data = [
    (datetime(2007,10,11),'Peak of bill market'),
    (datetime(2008,3,12),'Bear Sterarns Fails'),
    (datetime(2008,9,15),'Lehman Bankruptcy')
]
for date, label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
               xytext=(date, spx.asof(date)+225),
               arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
               horizontalalignment='left',verticalalignment='top')
    # ax.annotate 메서드를 이용해 x,y 좌표로 지정한 위치에 라벨을 추가
    
# 2007-2010 구간으로 확대
ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1800])         
    #set_xlim, set_ylim 메서드를 이용해 그래프의 시작과 끝 경게를 직접 지정했다

ax.set_title("Important dates in the 2008-2009 financial crisis")
```

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)

rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3) #좌측아래 꼭지점, 밑변, 높이. 각종 옵션
circ = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3)  #(중심좌표), 반지름, 각종 옵션
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5) #삼각형의 세 꼭지점

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```
