# pandas의 melt 함수
데이터프레임을 wide한 형태에서 long한 형태로 unpivot한다.

```python
import pandas as pd
df = pd.DataFrame({'A': {0: 'a', 1: 'b', 2: 'c'},
                   'B': {0: 1, 1: 3, 2: 5},
                   'C': {0: 2, 1: 4, 2: 6},
                  'D': {0:'kk',1:'gg',2:'jj'}})
```

```python
df
```

```python
df.melt(id_vars='A')
```

```python
df.melt(id_vars='A',value_vars='B')
```

```python
# list로 전달하는 것도 가능
df.melt(id_vars='A',value_vars=['B','C'])
```

```python
# 컬럼이름을 바꿀 수도 있다.
df.melt(id_vars='A',value_vars=['B','C'],var_name='그냥',value_name='이름')
```
