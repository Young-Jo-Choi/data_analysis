# 이동평균과 지수평균

```python
import zipfile
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import pandas as pd

# 아무 데이터나 가져와 사용
url = 'Bike-Sharing-Dataset.zip'
zf = zipfile.ZipFile(url)
df = pd.read_csv(zf.open('day.csv'))
```

```python
from matplotlib import font_manager,rc
import platform

if platform.system() == 'Windows':
    path = 'c:/windows/Fonts/malgun.ttf'
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font',family = font_name)
elif platform.system() == 'Darwin':
    rc('font',family = 'AppleGothic')
else:
    print('Check your OS system')
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
df['registered'].plot()
```

```python
# 시계열성이 존재한다고 여겨지는 부분을 슬라이스
df.iloc[350:600]['registered'].plot()
```

```python
data = df.iloc[350:600]['registered'].reset_index(drop=True)
```

```python
# window 크기 = 30
data.plot()
data.rolling(window=30).mean().plot()
plt.legend(['original','이동평균'])
```
