### pyplot.subplots 옵션

- nrows : 서브플롯의 수
- ncols : 서브플롯의 컬럼 수
- sharex : 모든 서브플롯이 같은 x축 눈금을 사용하도록 한다.(True or False)(xlim 값을 조절하면 모든 서브플롯에 적용됨)
- sharey : 모든 서브플롯이 같은 y축 눈금을 사용하도록 한다.(True or False)(ylim 값을 조절하면 모든 서브플롯에 적용됨)
- shubplot_kw : add_subplot을 사용해서 각 서브플롯을 생성할 때 사용할 키워드를 담고 있는 사전
- **fig_kw : figure을 생성할 때 사용할 추가적인 키워드 인자. ex) plt.subplots(2,2,figsize=(8,6))

## 서브플롯 간의 간격 조절하기
- subplots_adjust 메서드를 이용해서 바꿀 수 있음

plt.subplots_adjust(left=None,bottom=None, right=None, top=None,
                            wspace=None, hscape=None)
                            
- wspace, hspace는 서브플롯 간의 간격을 위해 각각 figure의 너비와 높이에 대한 비율을 조절

```python
# 다음은 서브플롯 간의 간격을 주지 않는 그래프

fig, axes = plt.subplots(2,2,sharex=True,sharey=True)  # 모든 서브플롯이 같은 x,y축 눈금을 사용하게 한다.

for i in range(2):
    for j in range(2):
        axes[i,j].hist(np.random.randn(500),bins=50,color='k',alpha=0.5)
        
plt.subplots_adjust(wspace=0,hspace=0)
```

### ax.plot(x,y,'g--')는 색상과 선 스타일을 지정하는 방법
### ax.plot(x,y,linestyle=='--',color='g')로 좀 더 명시적인 방법으로 가능함
아래는 그래프를 그리는 다양한 방법들


```python
fig, axes = plt.subplots(2,2,sharex=True,sharey=True)
axes[0,0].plot(np.random.randn(30).cumsum(),'ko--')
axes[1,0].plot(np.random.randn(30).cumsum(), color='g',linestyle = 'dashed',marker='*')

data = np.random.randn(30).cumsum()
axes[0,1].plot(data, color='b',linestyle = 'dashed',marker='d',label='Default')
axes[0,1].plot(data, color='r',label='steps-post', drawstyle='steps-post')
axes[0,1].legend(loc='best')            # 범례가 가장 적절한 위치에 추가됨(location을 best로 설정)
```

## 눈금, 라벨(ticks : 눈금)
set_xticks, set_xticklabels (옵션: fontsize,rotation), set_xlabel, set_title, set 등등

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.plot(np.random.randn(1000).cumsum())

# x축의 눈금이 이 5개 값으로 변경됨
ticks = ax.set_xticks([0,250,500,750,1000])

# 같은 위치값 찾아 축 이름 변경, 축 이름의 fontsize와 회전
labels =ax.set_xticklabels(['one','two','three','four','five'],fontsize='small',rotation=30)

 # 제목
ax.set_title('My first matplotlib plot')

# x축의 라벨
ax.set_xlabel('Stage')

# y축의 경우도 똑같이 진행 가능
```

```python
# 서브플롯의 set 이라는 메서드 하나로 할 수도 있음

props = {'title': 'My second','ylabel':'random number'}
ax.set(**props)
```

## 주석추가하기
- 글자는 text, arrow, annotate 함수를 이용해 추가할 수 있다.

주석은 글자와 화살표를 함께 써서 그릴 수 있는데 야후!파이낸스에서 얻은 2007년부터 S&P 500 지수 데이터로 그래프 생성하고 2008-2009년 사이에 있었던 재정위기 중 중요날짜를 주석으로 추가해볼 것임

```python
from datetime import datetime

fig = plt.figure()
ax = fig.add_subplot(1,1,1)

data = pd.read_csv('spx.csv',index_col=0, parse_dates=True)      # 컬럼 두 개짜리 단순하지만 양이 많은 data
spx = data['SPX']

spx.plot(ax=ax, style='k-')         # 그래프 그릴 때 Series나 DataFrame에 바로 plot을 사용할 수 있다.

crisis_data = [
    (datetime(2007,10,11),'Peak of bill market'),
    (datetime(2008,3,12),'Bear Sterarns Fails'),
    (datetime(2008,9,15),'Lehman Bankruptcy')
]

# annotate를 통해서 x,y 좌표로 지정한 위치에 라벨을 추가, 화살표까지 설정함

for date, label in crisis_data:
    ax.annotate(label,xy=(date,spx.asof(date)+75),
               xytext=(date, spx.asof(date)+225),
               arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
               horizontalalignment='left',verticalalignment='top')

# 2007-2010 구간으로 확대
# set_xlim, set_ylim 메서드를 이용해 그래프의 시작과 끝 경게를 직접 지정했다.

ax.set_xlim(['1/1/2007','1/1/2011'])
ax.set_ylim([600,1800])

ax.set_title("Important dates in the 2008-2009 financial crisis")
```

## 도형 추가

```python
fig = plt.figure(); ax = fig.add_subplot(1,1,1)

# 사각형 : 좌측아래 꼭지점, 밑변, 높이. 각종 옵션
rect = plt.Rectangle((0.2,0.75),0.4,0.15,color='k',alpha=0.3) 

# 원 : (중심좌표), 반지름, 각종 옵션
circ = plt.Circle((0.7,0.2),0.15,color='b',alpha=0.3) 

#삼각형의 세 꼭지점
pgon = plt.Polygon([[0.15,0.15],[0.35,0.4],[0.2,0.6]],color='g',alpha=0.5) 

ax.add_patch(rect)
ax.add_patch(circ)
ax.add_patch(pgon)
```

# 그래프를 파일로 저장하기
figure는 plt.savefig 메서드를 이용해 파일로 저장할 수 있다. 이 메서드는 figure 객체의 인스턴스 메서드인 savefig와 동일

plt.savefig('figpath.svg') 와 같이 입력

plt.savefig('figpath.png',dpi=400,bbox_inches='tight')

-->dpi는 인치당 도트 해상도 조절, bbox_inches는 실제 figure 둘레의 공백을 잘라낸다(tight는 그래프간 최소공백)

savefig 메서드는 파일에 저장할 뿐만 아니라 BytesIO처럼 파일과 유사한 객체에 저장하는 것도 가능
- from io import BytesIO
- buffer = BytesIO()
- plt.savefig(buffer)
- plot_data = buffer.getvalue()

# matplotlib 설정

만일 figure의 크기를 10x10으로 전역설정하고 싶다면
- plt.rc('figure',figsize=(10,10))
    
rc 메서드의 첫번째 인자는 설정하고자 하는 figure, axes, xtick, ytick, grid, legend 및 다른 컴포넌트의 이름

    그 다음으로 설정할 값에 대한 키워드 인자를 넘기게 된다. 아래와 같이 dictionary 타입을 이용해도 된다.

font_options={'family':'monospace','weight':'bold','size':'small'}

plt.rc('font',**font_options)
