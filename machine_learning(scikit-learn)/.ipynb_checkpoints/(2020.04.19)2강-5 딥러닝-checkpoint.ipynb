## 다층 퍼셉트론(multilayer perceptions,MLP)

입력 : x[0],x[1],...<br>
계수 : w[0],w[1],...<br>
출력: y = 가중치 합

MLP에서는 가중치 합을 만드는 과정이 여러 번 반복된다. 중간 단계를 구성하는 은닉 유닛(hidden unit)을 계산하고 이를 이용해 최종 결과를 산출하기 위해 다시 가중치 합을 계산

이 모델은 많은 계수를 학습해야하는데 계수들은 각 은닉층마다 있다. 여러 개의 가중치 합을 계산해야하는 이 모델을 하나의 가중치 합을 계산하는 것보다 강력하게 하기 위해서는, 각 은닉 유닛의 가중치 합을 계산한 후 그 결과에 비선형 함수인 relu나 tanh를 적용한다.<br>이 함수 결과의 가중치 합을 계산하여 출력 y를 만든다.

```python
line = np.linspace(-3,3,100)
plt.plot(line, np.tanh(line),label='tanh')
plt.plot(line, np.maximum(line,0), linestyle='--',label='relu')
plt.legend(loc='best')
plt.xlabel('x')
plt.ylabel("relu(x), tanh(x)")
```

$h[0] = tanh(w[0,0]*x[0]+w[1,0]*x[1]+w[2,0]*x[2]+w[3,0]*x[3]+b[0]$
$h[1] = tanh(w[0,1]*x[0]+w[1,1]*x[1]+w[2,1]*x[2]+w[3,1]*x[3]+b[1]$
$h[2] = tanh(w[0,2]*x[0]+w[1,2]*x[1]+w[2,2]*x[2]+w[3,2]*x[3]+b[2]$<br>
$y_p = v[0]*h[0]+v[1]*h[1]+v[2]*h[2]$<br>
보통 분류의 경우 신경망 마지막 출력층에 시그모이드(이진 분류)나 소프트맥스(다중 분류) 함수를 적용해 최종 출력을 계산한다.

w는 입력 x와 은닉층 h 사이의 가중치이고,v는 은닉층 h와 출력 $y_p$ 사이의 가중치이다.<br>
v,w는 훈련데이터로 학습하고, x는 입력 특성이며. $y_p$는 계산된 출력, h는 중간 계산값이다.

은닉층은 추가될 수 있으며 많은 은닉층으로 구성된 대규모 신경망이 생기며 이를 딥러닝이라 부른다.

## 신경망 튜닝

```python
from sklearn.neural_network import MLPClassifier
from sklearn.datasets import make_moons
```

```python
X,y = make_moons(n_samples=100, noise=0.25, random_state=3)
X_train, X_test, y_train,y_test = train_test_split(X,y,stratify=y,random_state=42)
```

```python
mlp = MLPClassifier(solver='lbfgs', random_state=0).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

MLP는 기본값으로 은닉 유닛 100개를 사용하는데 이런 작은 데이터셋에는 지나친 크기이다. 은닉 유닛 개수를 줄여도 여전히 좋은 결과를 얻으면서 모델의 복잡도를 낮출 수 있다.

```python
# 10개의 유닛으로 된 1개의 은닉층
mlp = MLPClassifier(solver='lbfgs', random_state=0,hidden_layer_sizes=[10]).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

기본 비선형 함수는 relu인데 은닉층이 하나이므로 결정 경계를 만드는 함수는 직선 10개가 합쳐져서 구성된다.<br>
더 매끄러운 경계를 원한다면 은닉 유닛을 추가하거나, 은닉층을 추가하거나 혹은 tanh함수를 사용할 수 있다.

```python
# 10개의 유닛으로 된 두 개의 은닉층
mlp = MLPClassifier(solver='lbfgs', random_state=0,hidden_layer_sizes=[10,10]).fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

```python
# 10개,11개의 유닛으로 된 두 개의 은닉층(두 은닉층의 유닛이 달라도 상관없다)과 tanh 함수가 적용
mlp = MLPClassifier(solver='lbfgs', random_state=0,hidden_layer_sizes=[10,11],activation='tanh').fit(X_train,y_train)
mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=0.3)
mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

리지 회귀와 선형 분류기에서 한 것처럼 L2 페널티를 사용해 가중치를 0에 가깝게 감소시켜 모델의 복잡도를 제어할 수 있다.<br>MLPClassifier에서 이런 역할을 하는 매개변수는 alpha(선형 회귀 모델과 같다)이고 기본값은 규제를 거의 하지않도록 매우 낮게 되어 있다.

```python
# 10개와 100개로 된 두 개의 은닉층을 사용했을 때, 다른 alpha값이 미치는 영향을 보여준다.
fig,axes = plt.subplots(2,4,figsize=(20,8))
for axx,n_hidden_nodes in zip(axes,[10,100]):
    for ax, alpha in zip(axx,[0.0001,0.01,0.1,1]):
        mlp = MLPClassifier(solver='lbfgs', random_state=0,
                           hidden_layer_sizes=[n_hidden_nodes,n_hidden_nodes],
                           alpha=alpha)
        mlp.fit(X_train,y_train)
        mglearn.plots.plot_2d_separator(mlp,X_train,fill=True,alpha=.3,ax=ax)
        mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,ax=ax)
        ax.set_title("n_hidden = [{},{}]\nalpha={:.4f}".format(
        n_hidden_nodes,n_hidden_nodes,alpha))
```

```python
# 실제 데이터인 유밥암 데이터셋
from sklearn.datasets import load_breast_cancer
cancer = load_breast_cancer()
print('특성별 최댁값:\n',cancer.data.max(axis=0))
```

```python
X_train,X_test,y_train,y_test = train_test_split(
cancer.data,cancer.target,random_state=0)
```

```python
mlp = MLPClassifier(random_state=42)
mlp.fit(X_train,y_train)
print(mlp.score(X_train,y_train))
print(mlp.score(X_test,y_test))
```

```python
# 데이터를 평균 0, 분산 1이 되도록 스케일링하는 것이 좋다.

mean_train = X_train.mean(axis=0)
std_train = X_train.std(axis=0)

X_train_scaled = (X_train-mean_train)/std_train

#(훈련데이터의 평균과 표준편차를 이용해)
X_test_scaled = (X_test-mean_train)/std_train

mlp = MLPClassifier(random_state=0)
mlp.fit(X_train_scaled,y_train)

print(mlp.score(X_train_scaled,y_train))
print(mlp.score(X_test_scaled,y_test))

# 아래 경고는 최대 반복 횟수에 도달했다는 것. adam 알고리즘에 관한 것으로 이 경고가 나오면 반복횟수를 늘려야 한다.
```

```python
mlp = MLPClassifier(max_iter=1000,random_state=0)
mlp.fit(X_train_scaled,y_train)

print(mlp.score(X_train_scaled,y_train))
print(mlp.score(X_test_scaled,y_test))
```

```python
# 일반화 성능을 높이도록 한다. alpha=0.0001 -> 1
mlp = MLPClassifier(max_iter=1000,random_state=0,alpha=1)
mlp.fit(X_train_scaled,y_train)

print(mlp.score(X_train_scaled,y_train))

# 데이터셋이 작거나 무리와 동떨어진 데이터 포인트가 있으면 어쩔 수 없이 조금씩 틀리게 됨
print(mlp.score(X_test_scaled,y_test))
```

### 참고) Adam(Adaptive Moment Estimation) : <br>
경사 하강법의 학습률인 learning rate 매개변수를 학습이 반복됨에 따라 조정해가는 알고리즘.<br>
MLPClassifier과 MLPRegressor의 solver 매개변수 기본값이 adam이고 max_iter 매개변수의 기본값은 200이다. <br>
solver가 sgd 또는 adam일 때 early_stopping을 True로 설정하면 연속된 검증 점수가 tol만큼 향상되지 않을 경우 반복을 자동으로 조기 종료시킬 수 있다.

```python
# 입력과 은닉층 사이에 학습된 가중치, 다음의 행은 30개의 입력 특성에 해당하며 열은 100개의 은늑 유닛에 해당한다.
plt.figure(figsize=(20,5))
plt.imshow(mlp.coefs_[0],interpolation='none',cmap='viridis')
plt.yticks(range(30),cancer.feature_names)
plt.xlabel("은닉 유닛")
plt.ylabel("입력 특성")
plt.colorbar()

# mlp.coefs_[0]은 입력과 은닉층 사이의 가중치가 저장되어 있는 (30,100)크기의 NumPy 배열
# mlp.coefs_[1]은 은닉층과 출력 사이의 가중치가 저장되어 있는 (100,1) 크기의 NumPy 배열
```

### 장단점과 매개변수
크고 강력한 모델이라면 종종 학습이 오래 걸린다.또한 데이터 전처리에 주의해야한다.SVM과 비슷하게 모든 틍성이 같은 의미를 가진 동질의 데이터에서 잘 작동한다. 다른 종류의 특성을 가진 데이터라면 트리기반 모델이 더 잘 동작할 수 있다.

각 은닉층의 유닛 수는 보통 입력 특성의 수와 비슷하게 설정하지만 수천 초중반을 넘는 일은 거의 없다.

신경망의 매개변수를 조정하는 일반적 방법은 먼저 충분히 과대적합되어 문제를 잘 해결할만한 큰 모델을 만든다. 그런 다음 훈련 데이터가 충분히 학습될 수 있다고 생각할 때 신경망 구조를 줄이거나 규제 강화를 위해 alpha값을 증가시켜 일반화 성능을 향상시킨다.

solver 
- adam : 대부분의 경우에 잘 작동하지만 데이터의 스케일에 민감
- lbfgs : 안정적이지만 규모가 큰 모델이나 대량의 데이터셋에서는 시간이 오래 걸린다.
- sgd : 다른 여러 매개변수와 함께 튜닝하여 최선의 결과를 만들 수 있다.

# 분류 예측의 불확실성 추정
어떤 데이터 포인트에 대해 분류기가 예측한 클래스가 무엇인지 뿐 아니라 정확한 클래스임을 얼마나 확신하는지가 중요한 때가 많다.<br>
어떤 질병을 예측할 때 거짓 양성은 환자에게 추가적인 진료를 요구하지만 거짓 음성은 심각한 질병을 치료하지 못하게 하는 결과를 만들게 된다.

scikit-learn 분류기에서 불확실성을 추정하는 함수는 decision_function과 predict_proba이다. 많은 종류의 분류 클래스는 적어도 둘 중 하나 이상의 함수를 제공한다.

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import make_circles
X,y = make_circles(noise=0.25, factor = 0.5, random_state=1)

# 예제를 위해 클래스의 이름을 blue와 red로 바꾼다.
y_named = np.array(['blue','red'])[y]

# 여러 개의 데이터셋을 한꺼번에 train_test_split에 넣을 수 있다.

X_train,X_test,y_train_named,y_test_named,y_train,y_test = \
train_test_split(X,y_named,y,random_state=0)

gbrt = GradientBoostingClassifier(random_state=0)
gbrt.fit(X_train,y_train_named)
```

## 결정 함수
이진 분류에서 decision_function 반환값의 크기는 (n_samples,)이며 각 샘플이 하나의 실수 값을 반환한다.

```python
X_test.shape
```

```python
gbrt.score(X_test,y_test_named)
```

```python
gbrt.decision_function(X_test).shape
```

이 값은 모델이 데이터 포인트가 양성 클래스인 클래스 1에 속한다고 믿는 정도이다. 양수 값은 양성 클래스를 의미하며 음수 값은 음성 클래스를 의미한다.

```python
gbrt.decision_function(X_test)
```

```python
# 임계치와 결정 함수 결과 비교
gbrt.decision_function(X_test)>0
```

```python
# 예측
gbrt.predict(X_test)
```

```python
# 불리언 값을 0,1로 변환
greater_zero = (gbrt.decision_function(X_test)>0).astype(int)

# classes_에 인덱스로 사용한다.
pred = gbrt.classes_[greater_zero]

# pred와 gbrt.predict의 결과를 비교한다.
np.all(pred == gbrt.predict(X_test))
```

```python
greater_zero
```

```python
pred
```

```python
# decision_function 값의 범위는 데이터와 모델 파라미터에 따라 달라진다.

decision_function = gbrt.decision_function(X_test)
print("최소값: {:.2f}, 최대값: {:.2f}".format(np.min(decision_function),np.max(decision_function)))
```

```python
fig,axes = plt.subplots(1,2,figsize=(13,5))

mglearn.tools.plot_2d_separator(gbrt,X,ax=axes[0],alpha=.4,
                               fill=True,cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(gbrt,X,ax=axes[1],alpha=.4,cm=mglearn.ReBl)
for ax in axes:
    # 훈련 포인트와 테스트 포인트를 그리기
    mglearn.discrete_scatter(X_test[:,0],X_test[:,1],y_test,
                            markers='^',ax=ax)
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,
                            markers='o',ax=ax)
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
cbar = plt.colorbar(scores_image,ax=axes.tolist())
cbar.set_alpha(1)
cbar.draw_all()
axes[0].legend(["테스트 클래스 0","테스트 클래스 1","훈련 클래스 0","훈련 클래스 1"],ncol=4,loc=(.1,1.1))
```

## 예측 확률
predict_proba의 출력은 각 클래스에 대한 확률이고 decision_function의 출력보다 이해하기 더 쉽다.<br>
이 값의 크기는 이진 분류에서 항상 (n_samples,2)이다.

```python
gbrt.predict_proba(X_test).shape
```

```python
# 각각 0일 확률, 1일 확률

gbrt.predict_proba(X_test)
```

출력값을 보면 분류기가 대부분의 포인트에서 비교적 강하게 확신하고 있다. 데이터에 있는 불확실성이 얼마나 이 값에 잘 반영되는지는 모델과 매개변수 설정에 달렸다. 과대적합된 모델은 잘못된 예측이더라도 예측의 확신이 강한 편.

일반적으로 복잡도가 낮은 모델은 예측에 불확실성이 더 많다. 이런 불확실성과 모델의 정확도가 동등하면 이 모델이 보정되었다고 한다.<br>즉 보정된 모델에서 70&확신을 가진 예측은 70%의 정확도를 낼 것이다.

```python
fig,axes = plt.subplots(1,2,figsize=(13,5))

mglearn.tools.plot_2d_separator(
gbrt,X,ax=axes[0],alpha=.4, fill=True,cm=mglearn.cm2)
scores_image = mglearn.tools.plot_2d_scores(
gbrt,X,ax=axes[1],alpha=.4, cm=mglearn.ReBl,function='predict_proba')

for ax in axes:
    # 훈련 포인트와 테스트 포인트 그리기
    mglearn.discrete_scatter(X_test[:,0],X_test[:,1],y_test,markers='^',ax=ax)
    mglearn.discrete_scatter(X_train[:,0],X_train[:,1],y_train,markers='o',ax=ax)
    ax.set_xlabel("특성 0")
    ax.set_ylabel("특성 1")
# colorbar를 감추지 않는다.
cbar = plt.colorbar(scores_image,ax=axes.tolist())
cbar.set_alpha(1)
cbar.draw_all()
axes[0].legend(["테스트 클래스 0",'테스트 클래스 1','훈련 클래스 0','훈련 클래스 1'],ncol=4,loc=(.1,1.1))
```

이 그래프의 경계는 훨씬 잘 나타나 있으며 불확실성이 있는 작은 영역들도 잘 확인할 수 있다.

# 다중 분류에서의 불확실성
클래스가 3개인 iris 데이터셋<br>
우선 decision_function

```python
from sklearn.datasets import load_iris
iris = load_iris()
X_train,X_test,y_train,y_test = train_test_split(iris.data, iris.target,random_state=42)
```

```python
gbrt = GradientBoostingClassifier(learning_rate=0.01,random_state=0)
gbrt.fit(X_train,y_train)
```

다중 분류에서 predict_proba와 decision_function의 결과값은 모두 (n_samples,n_classes)인데 이진분류의 decision_function만은 열이 하나 뿐이며 양성클래스에 대응하는 값만을 가지고 있다. (아마 오래전부터의 관례 때문인듯)

열이 n_classes개일 때는 열을 가로질러 argmax함수를 적용해 예측결과를 재현할 수 있다. 그러나 클래스가 문자열이거나 또는 정수형을 사용하지만 연속적이지 않고 0부터 시작하지 않을 수도 있다.predict의 결과와 예측함수의 결과를 비교하려면 분류기의 classes_속성을 사용해 클래스의 실제 이름을 얻어야 한다.

```python
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
named_target = iris.target_names[y_train]
logreg.fit(X_train,named_target)
print("\n훈련 데이터에 있는 클래스 종류 : ",logreg.classes_)
a=logreg.predict(X_test)[:10]
print("\n예측 : ",a)
argmax_dec_func = np.argmax(logreg.decision_function(X_test),axis=1)
print("\n가장 큰 결정 함수의 인덱스 : ",argmax_dec_func[:10])
b=logreg.classes_[argmax_dec_func][:10]
print("\n인덱스를 클래스에 연결 : ",b)
```

```python
np.all(a==b)
```
