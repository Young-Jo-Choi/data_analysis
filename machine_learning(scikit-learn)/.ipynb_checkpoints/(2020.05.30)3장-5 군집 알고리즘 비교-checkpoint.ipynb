# 군집 알고리즘의 비교와 평가
## 타깃값으로 군집 평가하기
최적일 때 1과 무작위로 분류될 때 0 사이의 값을 제공하는 ARI와 NMI라는 지표가 있다.(NMI는 음수가 될 수도 있음)

```python
from sklearn.datasets import make_moons
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.preprocessing import StandardScaler
from sklearn.metrics.cluster import adjusted_rand_score
```

```python
X,y = make_moons(n_samples=200,noise=0.05,random_state=0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

fig,axes = plt.subplots(1,4,figsize=(15,3),subplot_kw={'xticks':(),'yticks':()})

algorithms = [KMeans(n_clusters=2),AgglomerativeClustering(n_clusters=2),DBSCAN()]

# 비교를 위해 무작위로 클러스터를 할당
random_state = np.random.RandomState(seed=0)
random_clusters = random_state.randint(low=0,high=2,size=len(X))

axes[0].scatter(X_scaled[:,0],X_scaled[:,1],c=random_clusters,cmap=mglearn.cm3,s=60,edgecolors='black')
axes[0].set_title('무작위 할당 - ARI: {:.2f}'.format(adjusted_rand_score(y,random_clusters)))

for ax,algorithm in zip(axes[1:],algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    ax.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters,cmap=mglearn.cm3,s=60,edgecolors='black')
    ax.set_title("{} - ARI: {:.2f}".format(algorithm.__class__.__name__,adjusted_rand_score(y,clusters)))
```

참고) NMI를 위한 함수는 normalized_mutual_info_score이고 사용법은 adjusted_rand_score과 같다.

클러스터의 레이블 이름은 그 자체로 의미가 있는 것이 아니고 포인트들이 같은 클러스터에 속해 있는가만이 중요하므로 accuracy_score을 사용하는 것은 옳지 않다. 다음의 예를 통해서 확인할 수 있다.

```python
from sklearn.metrics import accuracy_score

cluster1 = [0,0,1,1,0]
cluster2 = [1,1,0,0,1]

print(accuracy_score(cluster1,cluster2))
print(adjusted_rand_score(cluster1,cluster2))
```

## 타깃값 없이 군집 평가
보통의 군집 알고리즘을 적용할 때에는 비교할 타깃값이 없기 때문에 ARI 같은 방법에는 제약이 있다. ARI나 NMI 같은 지표는 애플리케이션의 성능 평가가 아니라 알고리즘을 개발할 때나 도움이 된다.

타깃값 필요 없는 군집용 지표로는 실루엣 계수(silhouette coefficient)가 있다. 그러나 잘 동작하지 않을 수 있는데 이는 실루엣 점수가 클러스터의 밀집 정도를 계산하는 것이기 때문이다. 밀집된 클러스터가 좋긴 하지만 모양이 복잡할 때는 밀집도를 활용한 평가가 잘 들어맞지 않는다.(원형 클러스터에서 더 높게 나온다는 의미)

```python
from sklearn.metrics.cluster import silhouette_score

X,y = make_moons(n_samples=200,noise=0.05,random_state=0)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

fig,axes = plt.subplots(1,4,figsize=(15,3),subplot_kw={'xticks':(),'yticks':()})

# 비교를 위해 무작위로 클러스터를 할당
random_state = np.random.RandomState(seed=0)
random_clusters = random_state.randint(low=0,high=2,size=len(X))

axes[0].scatter(X_scaled[:,0],X_scaled[:,1],c=random_clusters,cmap=mglearn.cm3,s=60,edgecolors='black')
axes[0].set_title('무작위 할당 : {:.2f}'.format(silhouette_score(X_scaled,random_clusters)))

algorithms = [KMeans(n_clusters=2),AgglomerativeClustering(n_clusters=2),DBSCAN()]

for ax,algorithm in zip(axes[1:],algorithms):
    clusters = algorithm.fit_predict(X_scaled)
    ax.scatter(X_scaled[:,0],X_scaled[:,1],c=clusters,cmap=mglearn.cm3,s=60,edgecolors='black')
    ax.set_title("{} : {:.2f}".format(algorithm.__class__.__name__,silhouette_score(X_scaled,clusters)))
```

DBSCAN의 결과가 더 낫지만 k-평균의 실루엣 점수가 더 높다.

클러스터 평가에 더 적합한 지표는 견고성 기반(robustness-based)의 지표이다. 데이터에 잡음 포인트를 추가하거나 여러 가지 매개변수 설정으로 알고리즘을 실행하고 그 결과를 비교하는 것이다. 매개변수와 데이터에 변화를 주며 반복해도 결과가 일정하다면 신뢰할만 하다고 판단하는 방식이다.(scikit-learn에는 해당 기능이 구현되어있지 않다.)

## 얼굴 데이터셋으로 군집 알고리즘 비교

```python
from sklearn.datasets import fetch_lfw_people
```

```python
people = fetch_lfw_people(min_faces_per_person=20,resize=0.7)
```

```python
mask = np.zeros(people.target.shape,dtype=np.bool)   # 3023개 길이의 array
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1  # 각각의 target들을 같은 비중으로 최대 50개까지만 선택

X_people = people.data[mask]
y_people = people.target[mask]

# 0~255 사이의 흑백 이미지의 픽셀 값을 0~1 스케일로 조정
X_people = X_people/255
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=100,whiten=True,random_state=0)
X_pca = pca.fit_transform(X_people)
```

```python
X_pca.shape
```

### DBSCAN으로 얼굴 데이터셋 분석

```python
dbscan = DBSCAN()
labels = dbscan.fit_predict(X_pca)
print(np.unique(labels))
```

레이블이 -1뿐이므로 모든 데이터가 DBSCAN에 의해 잡음으로 레이블되었다. 그러므로 eps와 min_samples를 조정하기로 한다.

```python
dbscan = DBSCAN(min_samples=3,eps=15)
labels = dbscan.fit_predict(X_pca)
print(np.unique(labels))
```

```python
# 잡음 포인트와 클러스터에 속한 포인트의 수를 센다.
np.bincount(labels+1) #음수를 받을 수 없어 +1을 했음

# 32개는 잡음 포인트, 2031개는 클러스터에 속한 포인트의 수
```

```python
# 잡음 포인트는 많지 않으므로 확인 가능
noise = X_people[labels==-1]

image_shape = people.images[0].shape

fig,axes = plt.subplots(3,9,subplot_kw={'xticks':(),'yticks':()},figsize=(12,4))

for image,ax in zip(noise,axes.ravel()):
    ax.imshow(image.reshape(image_shape),vmin=0,vmax=1)
```

이것들은 DBSCAN에서 잡음으로 분류한 것들로써 손으로 입을 가린 이미지, 모자를 쓴 사람, 각도가 이상하거나 얼굴을 너무 가까이 혹은 멀리서 자른 경우들이다.

특이한 것을 찾아내는 이런 종류의 분석을 이상치 검출(outlier detection)이라고 한다. 실제 어플리케이션이라면 이미지 여백을 잘라 일정한 데이터셋을 만드는 것이 좋다. <br>
큰 클러스터 하나보다 더 많은 클러스터를 찾으려면 eps를 0.5(기본값)~15 사이 정도로 줄여야 한다.

```python
for eps in [1,3,5,7,9,11,13]:
    print('eps = ',eps)
    dbscan = DBSCAN(eps=eps,min_samples=3)
    labels = dbscan.fit_predict(X_pca)
    print("클러스터의 수 : ",len(np.unique(labels)))
    print("클러스터의 크기 : ",np.bincount(labels + 1))
```

eps가 작으면 모든 포인트가 잡음으로 분류된다. eps가 매우 크면 큰 클러스터 하나와 잡음 포인트를 만들어낸다.<br>
eps가 큰 경우 데이터에 확연히 구분되는 얼굴 이미지가 두세 개가 아니고 모든 이미지는 거의 동일하게 나머지 이미지들과 비슷하다는 것(혹은 비슷하지 않다는 것)을 의미한다.

작은 클러스터가 많이 만들어진 eps=7의 13개 클러스터의 포인들을 시각화해본다.

```python
dbscan = DBSCAN(min_samples=3,eps=7)
labels = dbscan.fit_predict(X_pca)

for cluster in range(max(labels)+1):
    mask = (labels == cluster)
    n_images = np.sum(mask)
    fig,axes = plt.subplots(1,14,figsize=(14*1.5,4),subplot_kw={'xticks':(),'yticks':()})
    
    i = 0
    for image,label,ax in zip(X_people[mask],y_people[mask],axes):
        ax.imshow(image.reshape(image_shape),vmin=0,vmax=1)
        ax.set_title(people.target_names[label].split()[-1])
        i+=1
    for j in range(len(axes)-i):
        axes[j+i].imshow(np.array([[1]*65]*87),vmin=0,vmax=1)
        axes[j+i].axis('off')
```

## k-means로 얼굴 데이터셋 분석하기
DBSCAN으로는 하나의 큰 클러스터 외에는 만들 수가 없다. 이에 비해 병합군집과 k-means는 비슷한 크기의 클러스터들을 만들 수는 있지만 클러스터의 개수를 지정해야만 한다. <br>
이 데이터셋에서는 실제 사람의 수를 알고 있으므로 클러스터의 수를 지정할 수 있지만, 비지도 군집 알고리즘이 원래대로 복원할 것 같지 않으므로 클러스터 개개를 분석해볼 수 있도록 클러스터 개수를 10개 정도로 줄여서 적용해보도록 한다.

```python
km = KMeans(n_clusters=10, random_state=0)
labels_km = km.fit_predict(X_pca)
print('k-평균 클러스터의 크기',np.bincount(labels_km))
```

DBSCAN과는 다르게 비교적 비슷한 크기인 91에서 358까지의 클러스터들로 나누었다.

k-평균 클러스터 중심을 시각화해보면 결과를 더 자세히 분석할 수 있다. PCA 성분으로 군집 알고리즘을 사용했기 때문에 k-means클러스터 중심을 pca.inverse_transform을 사용해 원본 공간으로 되돌린 후 시각화해야한다.

```python
fig,axes = plt.subplots(2,5,subplot_kw={'xticks':(),'yticks':()},figsize=(12,4))

for center,ax in zip(km.cluster_centers_,axes.ravel()):
    ax.imshow(pca.inverse_transform(center).reshape(image_shape),vmin=0,vmax=1)
```

```python
pca.inverse_transform(center).shape
```

```python
km.cluster_centers_.shape
```

위 이미지들은 각 클러스터의 중심(이 되는 이미지)들

다음은 각 클러스터에서 가장 대표되는 이미지(클러스터 중심 이미지에 가장 가까운 이미지) 다섯 개, 가장 동떨어진 이미지(클러스터 중심 이미지에서 가장 먼 이미지) 다섯 개

```python
mglearn.plots.plot_kmeans_faces(km,pca,X_pca,X_people,y_people,people.target_names)
```

## 병합 군집으로 얼굴 데이터셋 분석하기

```python
agglomerative = AgglomerativeClustering(n_clusters=10)
labels_agg = agglomerative.fit_predict(X_pca)
print("병합 군집의 클러스터 크기 : ",np.bincount(labels_agg))
```

```python
adjusted_rand_score(labels_agg,labels_km)
```

ARI 값이 0.1이라는 것은 두 군집 labels_agg와 labels_km 사이에 공통 부분이 거의 없다는 의미이다.

```python
from scipy.cluster.hierarchy import dendrogram, ward

linkage_array = ward(X_pca)

plt.figure(figsize=(20,5))
dendrogram(linkage_array, p=7, truncate_mode='level',no_labels=True)
plt.xlabel('샘플 번호')
plt.ylabel("클러스터 거리")
ax = plt.gca()
bounds = ax.get_xbound()
ax.plot(bounds,[36,36],'--',c='k')
```

k-means처럼 10개의 클러스터를 그림으로 나타낼 수 있다. 병합 군집에서는 클러스터 중심이라는 개념은 없으므로(물론 평균점을 계산할 수는 있다) 그냥 클러스터에 속한 몇 개의 포인트를 나타내보도록 하겠다.

```python
n_clusters=10
for cluster in range(n_clusters):
    mask = (labels_agg == cluster)
    fig,axes = plt.subplots(1,10,subplot_kw={'xticks':(),'yticks':()},figsize=(15,8))
    axes[0].set_ylabel(np.sum(mask))  # 1열의 왼쪽에는 각 클러스터에 속한 데이터 포인트의 개수
    for image,label,asdf,ax in zip(X_people[mask],y_people[mask],labels_agg[mask],axes):
        ax.imshow(image.reshape(image_shape),vmin=0,vmax=1)
        ax.set_title(people.target_names[label].split(' ')[-1],fontdict={'fontsize':9})
```

각 행은 하나의 클러스터에 해당

일부 클러스터들은 일관성을 가지기에는 너무 크기 때문에 클러스터의 수를 40개로 늘려 알고리즘을 실행한다.

```python
agglomerative = AgglomerativeClustering(n_clusters=40)
labels_agg = agglomerative.fit_predict(X_pca)
print("병합 군집의 클러스터 크기 : ",np.bincount(labels_agg))
```

```python
n_clusters = 40
for cluster in [13,16,23,38,39]: # 그냥 특별한 클러스터 몇개 골라낸거임
    mask = labels_agg == cluster
    fig,axes = plt.subplots(1,15,subplot_kw={'xticks':(),'yticks':()},figsize=(15,8))
    cluster_size = np.sum(mask)
    axes[0].set_ylabel("#{}: {}".format(cluster,cluster_size))
    for image,label,asdf,ax in zip(X_people[mask],y_people[mask],labels_agg[mask],axes):
        ax.imshow(image.reshape(image_shape),vmin=0,vmax=1)
        ax.set_title(people.target_names[label].split(" ")[-1],fontdict={'fontsize':9})
        
    for i in range(cluster_size,15):
        axes[i].set_visible(False)
```
