# k-means 군집
데이터의 어떤 영역을 대표하는 클러스터 중심(cluster center)을 찾는다. 처음 클러스터 중심은 무작위로 선정한다.

1. 데이터 포인트를 가장 가까운 클러스터 중심에 할당
2. 클러스터에 할당된 데이터 포인트의 평균으로 클러스터 중심을 다시 지정


- 반복 후 클러스터에 할당되는 데이터 포인트에 변화가 없을 때 알고리즘이 종료

```python
# 삼각형은 클러스터 중심
mglearn.plots.plot_kmeans_algorithm()
```

```python
# 위 학습의 클러스터 중심의 경계
mglearn.plots.plot_kmeans_boundaries()
```

```python
# KMeans 객체를 생성하고 찾고자 하는 클러스터의 수를 지정한다.(기본값은 8, 보통은 기본값 그대로 쓰지 않는다.)

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

X,y = make_blobs(random_state=1)

kmeans = KMeans(n_clusters=3)

# 알고리즘을 적용하면 X에 담긴 각 훈련 데이터 포인트에 클러스터 레이블이 할당된다.
kmeans.fit(X)
```

```python
# 알고리즘을 적용하면 X에 담긴 각 훈련 데이터 포인트에 클러스터 레이블이 할당된다. 
# 3개의 클러스터를 지정했기 때문에 0~2까지의 번호가 붙는다.
kmeans.labels_
```

```python
# 예측은 각 포인트에 가장 가까운 클러스터 중심을 할당, 훈련 세트에 대해 predict하면 labels_와 같은 결과를 얻는다.
kmeans.predict(X)
```

분류와 비슷해보이지만 정답을 모르고 있으며 레이블 자체에 어떤 의미가 있지 않다는 차이점이 존재한다.<br>
알고리즘이 우리에게 주는 정보라고는 같은 레이블을 가진 샘플들끼리 비슷하다는 것 정도이다.(0,1,2이냐 2,0,1이냐는 중요하지 않음)

```python
# 클러스터 수를 늘이거나 줄여보도록 한다,

fig,axes = plt.subplots(1,2,figsize=(10,5))

kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
assignments = kmeans.labels_
mglearn.discrete_scatter(X[:,0],X[:,1],assignments,ax=axes[0])

kmeans = KMeans(n_clusters=5)
kmeans.fit(X)
assignments = kmeans.labels_
mglearn.discrete_scatter(X[:,0],X[:,1],assignments,ax=axes[1])
```

## k-means 알고리즘이 실패하는 경우
데이터셋의 클러스터 개수를 정확히 알더라도 k-평균 알고리즘이 이를 항상 정확히 구분해낼 수는 없다.<br>
각 클러스터를 정희하는 것은 중심 하나뿐이므로 클러스터는 항상 둥근 형태로 나타난다.<br>
또한 k-평균은 모든 클러스터의 반경이 똑같다고 가정한다. 그래서 클러스터의 중심 사이의 정확히 중간에 경계를 그린다.

```python
X_varied,y_varied = make_blobs(n_samples=200,
                              cluster_std = [1.0,2.5,0.5],
                              random_state=170)
y_pred = KMeans(n_clusters=3,random_state=0).fit_predict(X_varied)
mglearn.discrete_scatter(X_varied[:,0],X_varied[:,1],y_pred)
plt.legend(['클러스터 0','클러스터 1','클러스터 2'],loc='best')
plt.xlabel("특성 0")
plt.ylabel('특성 1')

# 클러스터 0과 1인 비교적 중심에서 멀리 떨어진 포인트들도 포함한다.
```

k-평균은 클러스터에서 모든 방향이 똑같이 중요하다고 가정한다. 

```python
X,y = make_blobs(random_state=170,n_samples=600)
rng = np.random.RandomState(74)

# 데이터가 길게 늘어지도록 변경
transformation = rng.normal(size=(2,2))
X = np.dot(X,transformation)

# 세 개의 클러스터로 데이터에 KMeans 알고리즘을 적용
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# 클러스터 할당과 클러스터 중심을 나타냄
plt.scatter(X[:,0],X[:,1],c = kmeans.labels_,marker='o',cmap='coolwarm')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='^',color='black',linewidths=5)
plt.xlabel("특성 0")
plt.ylabel('특성 1')
```

```python
# 참고
kmeans.cluster_centers_.shape
```

클러스터 중심에서 가장 가까운 중심까지의 거리만 고려하기 때문에 이런 데이터를 잘 처리하지 못한다.

클러스터가 two_moons 데이터처럼 더 복잡한 형태라면 k-평균의 성능이 더 나빠진다.

```python
from sklearn.datasets import make_moons
X,y = make_moons(n_samples=200,noise=0.05, random_state=0)

# 두 개의 클러스터로 데이터에 KMeans 알고리즘을 적용한다.
kmeans = KMeans(n_clusters=2)
kmeans.fit(X)
y_pred = kmeans.predict(X)

plt.scatter(X[:,0],X[:,1],c=y_pred,cmap=mglearn.cm2,s=60,edgecolors='k')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],
           marker='^',c=[mglearn.cm2(0),mglearn.cm2(1)],s=100,linewidth=2,edgecolors='k')
plt.xlabel('특성 0')
plt.ylabel('특성 1')
```

군집 알고리즘이 반달모양을 구분하면 좋겠지만 k-평균 알고리즘으로는 어려움

## 벡터 양자화 또는 분해 메서드로서의 k-means
k-평균은 클러스터 중심으로 각 데이터 포인트를 표현한다. 이를 각 데이터 포인트가 클러스터 중심, 즉 하나의 성분으로 표현된다고 볼 수 있다. k-평균을 이렇게 각 포인트가 하나의 성분으로 분해되는 관점으로 보은 것을 벡터 양자화(vector quantization)라고 한다.


```python
# PCA,NMF,k-means에서 추출한 성분과 100개의 성분으로 테스트 세트의 얼굴을 재구성한 것을 비교
# k-means의 경우 재구성은 훈련 세트에서 찾은 가장 가까운 클러스터의 중심이다.

from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_lfw_people
people = fetch_lfw_people(min_faces_per_person=20,resize=0.7)

mask = np.zeros(people.target.shape,dtype=np.bool)   # 3023개 길이의 array
for target in np.unique(people.target):
    mask[np.where(people.target == target)[0][:50]] = 1  # 각각의 target들을 같은 비중으로 최대 50개까지만 선택
    
X_people = people.data[mask]
y_people = people.target[mask]
```

```python
from sklearn.decomposition import PCA,NMF
```

```python
X_train,X_test,y_train,y_test = train_test_split(X_people,y_people,random_state=42)
```

```python
pca = PCA(n_components=100,random_state=0)
pca.fit(X_train)
X_pca = pca.inverse_transform(pca.transform(X_test))
```

```python
nmf = NMF(n_components=100,random_state=0)
nmf.fit(X_train)
X_nmf = np.dot(nmf.transform(X_test),nmf.components_)
```

```python
kmeans = KMeans(n_clusters=100,random_state=0)
```

```python
kmeans.fit(X_train)
```

```python
X_kmeans = kmeans.cluster_centers_[kmeans.predict(X_test)]
```

```python
print(kmeans.cluster_centers_.shape)   # 100개의 중심점
print(kmeans.predict(X_test).shape)    
print(X_kmeans.shape)                  # 516개의 각각 그 중심점. nunique하게 만들면 100개이하 임
print(len(np.unique(X_kmeans,axis=0)))
```

```python
image_shape = people.images[0].shape


# 추출한 성분

fig,axes = plt.subplots(3,5,subplot_kw={'xticks':(),'yticks':()},figsize=(8,8))
fig.suptitle('추출한 성분')

# ax가 열별로 반복문 작동하도록
for ax, comp_kmeans, comp_pca,comp_nmf in zip(axes.T,kmeans.cluster_centers_,pca.components_,nmf.components_):
    ax[0].imshow(comp_kmeans.reshape(image_shape))
    ax[1].imshow(comp_pca.reshape(image_shape),cmap='viridis')
    ax[2].imshow(comp_nmf.reshape(image_shape))
    
axes[0,0].set_ylabel("kmeans")
axes[1,0].set_ylabel('pca')
axes[2,0].set_ylabel('nmf')
```

```python
# 재구성

fig, axes = plt.subplots(4,5,subplot_kw={'xticks':(),'yticks':()},figsize=(8,8))
fig.suptitle("재구성")
for ax,orig,rec_kmeans,rec_pca,rec_nmf in zip(axes.T,X_test,X_kmeans,X_pca,X_nmf):
    ax[0].imshow(orig.reshape(image_shape))
    ax[1].imshow(rec_kmeans.reshape(image_shape))
    ax[2].imshow(rec_pca.reshape(image_shape))
    ax[3].imshow(rec_nmf.reshape(image_shape))
    
axes[0,0].set_ylabel("원본")    
axes[1,0].set_ylabel("kmeans")
axes[2,0].set_ylabel('pca')
axes[3,0].set_ylabel('nmf')
```

k-평균을 사용한 벡터 양자화는 입력 데이터의 차원보다 더 많은 클러스터를 사용해 데이터를 인코딩할 수 있다.

```python
# two_moons 데이터는 PCA나 NMF를 사용해 1차원으로 축소하면 구조가 완전히 파괴된다. 
# 하지만 많은 클러스터 중심을 이용한 k-means는 데이터를 더 잘 표현할 수 있다,

X,y = make_moons(n_samples=200,noise=0.05,random_state=0)

# 10개의 중심점 주목
kmeans = KMeans(n_clusters=10,random_state=0)
kmeans.fit(X)
y_pred = kmeans.predict(X)

plt.scatter(X[:,0],X[:,1],c=y_pred,s=60,cmap='Paired',edgecolors='black')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],s=60,
           marker='^',c=range(kmeans.n_clusters),linewidth=2,cmap='Paired',edgecolors='black')
plt.xlabel("특성 0")
plt.ylabel('특성 1')
```

```python
print("클러스터 레이블 : \n",y_pred)
```

원래 특성 2개만으로는 불가능했지만, 10차원 형태를 이용하면 선형모델을 사용해 두 개의 반달 모양을 구분할 수 있다.

클러스터 중심까지의 거리를 특성으로 사용하면 데이터를 더욱 잘 나타낼 수 있다.

```python
distance_features = kmeans.transform(X)
print("형태 : ",distance_features.shape)
print("클러스터 거리 :\n",distance_features)
```

scikit-learn은 k-means 방법의 무작위 초기화에 대해 서로 다른 난수 값으로 10번 반복해 최선의 결과를 만든다. 최선의 결과라함은 클러스터 분산의 합이 작은 것을 의미한다.

k-means는 이해도 쉽고 구현도 쉬우면서 비교적 빠르고, 대용량 데이터셋에도 잘 동작한다.(MiniBatchKMeans라는 아주 큰 대규모 데이터셋에 대한 알고리즘 역시 scikit-learn에서 제공됨)<br>
그러나 클러스터의 모양이 원형으로 가정되어있고 찾으려는 클러스터의 개수를 지정해야하는데 실제 어플리케이션에서는 알기 힘든 부분이다.
